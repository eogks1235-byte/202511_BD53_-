# 파일 쓰기(file write)

1. 파일 열기(open)
1. 파일 쓰기(write)
1. 파일 닫기(close)

`open(file, mode, ...)` 함수:
*   파라미터 file: 폴더(디렉토리) 이름을 포한함 전체 파일 경로(이름).
*   파라미터 mode:
    *   `r`: read. 읽기 전용 모드. 기본값(생략 가능).
    *   `w`: write. 쓰기 모드. 기존에 작성된 내용을 모두 지우고 처음부터 새로 쓰기 시작.
    *   `a`: append. 추가 모드. 기존에 작성된 내용 뒤에 이어서 쓰기.
    *   `t`: text. 텍스트 읽기/쓰기. 기본값.
    *   `b`: binary. 바이너리(2진) 형식 파일(이미지, 동영상, ...) 읽기/쓰기.



```python
# 1. 파일 열기               r>>t
f= open(file='hello.txt',mode='w') #mode='wt'

# 2. 파일에 내용을 쓰기
f.write('안녕하세요.\n')
f.write('오늘은 목요일입니다.\n')
f.write('Hello, Python!\n')
f.write('2025-12-18\n')
# 3. 파일 닫기.
f.close()
```

파일을 열 때, mode 파라미터를 'w' 또는 'a'로 설정한 경우:
*   파일이 없으면 새로운 파일을 만듦.
*   파일이 있으면 기존 파일을 열어줌.
    *   'w' 모드는 기존 파일의 내용을 삭제.
    *   'a' 모드는 기존 파일의 내용이 유지.
mode 파라미터를 'r'로 설정한 경우에는 파일이 없으면 에러가 발생함!    

`with open(file, mode,...) as 변수:` 구문은 변수에서 close() 메서드를 자동으로 호출.

```python
with open(file='hello.txt',mode='w') as f:
  f.write('Hello...\n')
  f.write('안녕???\n')
  # f.close()는 자동으로 호출됨.
```

# 파일 읽기

1. 파일 열기(open)
1. 파일에서 내용을 읽기(read,readline,...)
1. 파일 닫기(close)

`with ... as ...` 구문을 사용하면 `close()` 메서드는 자동으로 호출됨.

```python
# with open('test.txt') as f:
#   pass
with open(file='hello.txt', mode='r') as f:
                            #mode='rt'와 같음. 생략가능(기본값).
  line = f.readline()   # 한 줄씩 읽기(줄바꿈(\n)까지 읽기)
  print(line.strip())   # strip() 메서드: 문자열의 시작/끝의 공백 문자들을 제거
  line = f.readline()
  print(line.strip())    #txt에 줄바꿈을 넣어서 +print(end=) =줄바꿈*2
```

`for var... in iterable:` 구문에서 파일 스트림 객체를 사용할 수 있음.

for 반복문에서는 `readline()` 메서드를 호출할 필요가 없음. 반복문이 한 줄씩 읽어줌.

```python
with open(file='hello.txt') as f:
  for line in f: # 파일 처음부터 끝까지 한 줄씩 읽어줌.
    print(line.strip())
```

class 작성연습


```python
class Student:
  def __init__(self,id,cls,math,english,science): #cls인 이유 클래스랑곂침
    self.id=int(id)
    self.cls=int(cls)
    self.math=int(math)
    self.english=int(english)
    self.science=int(science)

  def __repr__(self):
    return f'Student(아디={self.id},반={self.cls},수학={self.math},영어={self.english},과학={self.science})'
  # def mean(self):
  #   mean=sum(self.math)/len(self.math)
  #   return  mean
```

```python
# 파일에서 읽은 내용을 Student객체로 만들어서 리스트에 저장.
exam_student=[]

with open(file=file_path) as f:
  f.readline()
  for line in f:
    row= line.strip().split(',')
    student= Student(row[0],row[1],row[2],row[3],row[4])
    exam_student.append(student)
exam_student
```
