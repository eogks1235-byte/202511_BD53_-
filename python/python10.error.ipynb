코드 실행 중에 오류가 발생하는 경우 해결 방법:

*   문법적인 오류가 아니라 예외적인 상황이 발생할수 있는 코드인 경우에는 **예외 처리**를 해야함


# try-except 구문

```
try:
    (1) 실행할 코드 블록
except [에러 이름 [as 별명]]:
    (2) 예외 상황이 발생했을 때 예외를 처리하는 코드 블록
[
else:
    (3) try 블록의 모든 코드가 예외 없이 정상 수행됐을 때 실행할 코드 블록
finally:
    (4) 예외 발생 여부 상관 없이 항상 실행할 코드 블록    
]        
```

* try 구문의 실행 순서
  *    예외가 발생한 경우: (1) -> (2) -> (4)
  *    예외가 발생하지 않은 경우: (1) -> (3) -> (4)
* except 절에서 에러 이름을 생략하면 모든 종류의 에러들을 처리할 수 있다
* except 절에서 에러 이름을 지정하면, 선언된 에러만 처리할 수 있다  
* 처리해야 할 에러의 종류가 여러 개인 경우에는 except 절을 여러 개 사용할 수 있다

````python
try:
  x=int(input('입력 >>> ')) #ValueError
  y=int(input('입력 >>> '))

  print(f'{x}+{y}= {x+y}')
  print(f'{x}-{y}= {x-y}')
  print(f'{x}x{y}= {x*y}')
  print(f'{x}/{y}= {x/y}') #ZeroDivisionError 별명이름 중복 가능 , print가능
except ValueError as ve:
  print('x와 y는 정수로 입력해야 합니다')
  print(ve)
except ZeroDivisionError as zde:
  print('0은안놰')
  print(zde) #실제에러메세지
else:
  print('사칙연산을 모두 수행 했습니다')
finally:
  print('코드 셀 종료')  ## try 예외 except 1  2 finally
````

````python
#ValueError, IndexError, 오류 사항 전부다 포함이다
try:
  index = int(input('인덱스 입력 >> '))
  print(f'numbers[{index}] = {numbers[index]}')# 파이썬 코드
except:
  print('잘못된 인덱스를 입력했습니다')
````
