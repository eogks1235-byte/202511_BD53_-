# `str` 클래스 메서드

```python
s= 'heLLo, pYThon!, HeLlo Python. 안녕하세요?' #일반적인방법
#s=str() 생성자호출
```

```python
type(s)#type(arg):arg의 타입(클래스 이름)을 리턴
```

```python
print(s) #변수 s의 __str__메서드가 리턴하는 문자열을 출력
```

```python
s # 변수 s의 __repr__메서드가 리턴하는 문자열을 출력
```

```python
result= s.lower() # 문자열 s를 소문자로 변환한 "새로운" 문자열을 리턴. s는 변경되지 않음
print(result)
print(s)
```

```python
result=s.upper() #문자열 s를 대문자로 변환한 "새로운" 문자열 리턴
print(result)
print(s)
```

```python
s.capitalize() #해당 문자열에서 첫글자만 대문자로 변환해서 출력
```

```python
s.count('h') # 모든값들은 리턴값 출력 문자열 내에서 소문자h의 갯수
```

```python
# 대/소문자 구분없이 h가 몇개있는지?
# result=s.upper(s.count('h'))
# print(result)
print(s.lower().count('h'))
#s.count('h').lower()
# s.upper()
# s.upper().count
```

```python
email='gildong@itwill.com'

# 문자열 email에서 아이디(gildong)와 도메인(itwill.com)을 구분해서 출력
result=email[0:7]
print(result)

idx=email.find('@') # 문자열 email에서 '@'가 있는 인덱스를 리턴
print(idx)
print(email[:idx])  #문자열slicing
print(email[(idx +1):]) #idx는 골뱅이니까
```

```python
msg='파이썬 쉽죠? 열심히 공부합시다!'
result=msg.split()
#문자열저장리스트 , 문자열msg를 공백(whitespace: space, tab, new line)을 기준으로 자른 문자열들의 리스트를 리턴
print(result)
```

```python
date_str='2025-12-17'
result=date_str.split('-')
#문자열 date_str을 '-'를 기준으로 자른 문자열들의 리스트를 리턴
print(result)
```

```python
print(email)
result=email.split('@')
print(result)
print(result[0])
print(result[1])
```

```python
# csv(comma-separated values)
csv_str='1, 홍길동,16,학생'
csv_str.split(',')
```

```python
sentence='    Hello, Python!       안녕하세요.       '
print(sentence)
print(sentence.lstrip())
print(sentence.rstrip())
print(sentence.strip())#웹크롤링
```

```python
s= 'hello python'
s.replace('h','H').replace('o','O')
# 문자열 s에서 'h'를 'H'로 변환한 문자열을리턴
```

```python
print(s)
s.replace('h','H',1) #1개만(갯수) -1은 전부(기본값)
```
