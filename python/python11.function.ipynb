*   **함수(Function)**: 기능을 수행하는 코드 블록.
*   **인수(argument)**: 함수를 호출할 때 함수에게 전달하는 값.
*   **매개변수(parameter)**: 아규먼트를 저장하기 위해서 함수를 정의하는 곳에서 선언하는 변수.(인자)

````python
#
def make_list(start, end, n):

  """
  @param start : 시작점 ,int
  @param end : 미만값 , int
  @param n : 갯수 ,0또는 양의정수
  @return : 정수난수 n개의 리스트

  """
  # lists=[random.randrange(start,end) for x in range(n)]
  # return lists

  numbers=[]
  for _ in range(n):
    numbers.append(random.randrange(start,end))

  return numbers

````

함수를 호출할 때 아규먼트를 전달하는 방법:

*   **positional argument**:  선언된 파라미터 순서대로 아규먼트들을 전달하는 방식

      * 키워드
      * keyword argument 방식은 파라미터 순서를 지키지 않아도 됨
      * **(주의)** 함수를 호출할 때 positional argument와 keyword argument를 함께 사용하는 경우에는, 반드시 positional argument들을 먼저 써야함

# 가변길이 인수(variable-length arguments)

*   함수를 호출할 때 아규먼트 몇 개를 전달해도 됨
    *    아규먼트를 한 개도 전달하지 않아도 됨
*   함수를 선언할 때 파라미터 이름 앞에 `*`를 사용한 파라미터가 가변길이 인수가 됨
*   함수 내부에서 가변길이 인수는 튜플(tuple)로 취급함
    *     가변길이 인수는 인덱스를 가짐
    *     가변길이 인수는 `for-in` 구문에서 사용할 수 있음
*   함수를 선언할때 가변길이 인수는 1개만 선언할 수 있음
*   가변길이 인수를 keyword argument 방식으로 아규먼트들을 전달할 수 없음
*   가변길이 인수 뒤에서 선언된 파라미터(들)은 keyword argument 방식으로만 아규먼트를 전달할 수 있다

# 가변길이 키워드 아큐먼트(variable-length keyword argument)

*   함수를 선언할 때 파라미터 이름 앞에 `**`를 사용한다
*   함수를 호출할 때, `param1=value1, param2=value2, ...` 형식으로 아규먼트들을 전달하면 됨
      * 가변길이 아규먼트--> 전달하는 아규먼트 개수는 제한이 없음
      * 키워드 아규먼트----> 반드시 키워드(파라미터 이름)와 함께 아규먼트들을 전달해야 함
          * 키워드는 자유롭게 만듦
*   함수 내부에서 가변길이 키워드 아규먼트는 dict 타입 객체로 취급
      * 키워드(파라미터 이름)는 dict의 key가 됨
      * 아규먼트는 dict의 value가 됨
*   가변길이 키워드 아규먼트는 함수에서 오직 1개만 선언할 수 있음

````python
def make_student_info(id, name, **kwargs):
  student ={
      'id' : id,
      'name' : name
  }
  print(f'kwargs = {kwargs}')
  for k,v in kwargs.items():
    student[k]=v

  return student
````
