## 내부 함수, 함수를 리턴하는 함수

내부 함수(inner function), 지역 함수(local function):
*   함수 내부에서 선언된 함수.
*   선언된  함수(외부 함수) 안에서만 호출할 수 있음.
*   선언된 함수 바깥에서는 호출할 수 없음.
*   외부 함수의 지역 변수들(파라미터 포함)을 사용할 수 있음.


````python
def make_incromentor(n):
  # 내부 함수 선언
  def add_n(x):
    return x+n #안쪽리턴 add_n(2)X

  #함수객체 리턴
  return add_n  #최종리턴
````

````python
add_n(100) #> 내부(지역) 함수는 단독으로 사용할 수 없음
````

````python
plus_2=make_incromentor(2)
print(plus_2)
````

````python
plus_2(100)
````

````python
plus_2=make_incromentor(10) # 1. 10 값에 대한 함수 리턴(add_n)=x+10  2. add_n(100)= 100+10
plus_2(100)
````

````python
make_incromentor(3)(100) #함수 재호출
````

## lambda를 아규먼트로 전달하는 예
#람다가 함수가 어떻게 생긴지 몰라도 사용을 한다는게 무슨소리임? 간단식은 만들수 있다는 소리였음

*   filter: 조건을 만족하는 아이템들을 찾기. `filter (function, iterable)`
    *   iterable 안에 있는 원소들을 하나씩 순서대로 function의 아규먼트로 전달하면서 function을 호출하고, function이 True를 리턴하는 원소들만 선택한다
    *   function은 아규먼트가 하나다 boolean타입 True, False를 리턴하는 함수
    *   `lambda x: True/False를 리턴하는 식`
*   mapping: 아이템들을 일정한 규칙에 따라서 다른 값으로 변환(매핑). `map (function, iterable)`
    *   iterable 안에 있는 원소들을 하나씩 순서대로 function의 아규먼트로 전달하면서 function을 호출하고, function이 리턴하는 값으로 매핑 값을 만듦.
    *   function은 아규먼트가 1개이고, 리턴값이 있는 함수.
    *   `lambda x: 리턴값(식)`

````python
import random
````

````python
numbers=[random.randrange(100) for x in range(10)]
numbers
````

````python
# numbers의 원소들 중에서 짝수들만 필터링
[x for x in numbers if x%2==0]
````

## filter 함수 구현

````python
def my_filter(iterable, func):
  """
  필터링. iterable의 원소들 중에서 func의 아규먼트로 전달했을 때 func의 리턴값이 True가 되는 원소들로 이루어진 리스트를 리턴.

  @param iterable - list 또는 tuple
  @param func - 아규먼트가 1개이고 True 또는 False를 리턴하는 함수.
  @return 필터링 결과를 저장한 리스트(list)
  """

  result =[] # 필터링한(조건을 만족하는) 아이템들을 저장하기 위한 리스트.
  for x in iterable:  # iterable의 원소들을 처음부터 끝까지 순서대로 반복
    if func(x):   # iterable의 원소를 func의 아규먼트로 전달 했을 때 True가 되면
      result.append(x) # x를 리스트에 추가한다
  #result=[x for x in iterable if fun(x)]

  return result
````

````python
numbers= [random.randrange(-10,11) for x in range(10)]
print(numbers)
````

````python
# numbers에서 짝수들만 필터링
my_filter(numbers,lambda x: x%2==0)
````

````python
# numbers에서 양수들만 필터링
my_filter(numbers, lambda x: x>0)
````

````python
languages= ['SQL', 'Python', 'R', 'JavaScript']

#languages에서 5글자 이상인 문자열들만 필터링
my_filter(languages, lambda x: len(x)>=5)
````

````python
def my_map(iterable, func):

  """
  매핑. iterable의 원소를 func의 아규먼트로 전달해서 호출했을 때 그 리턴값으로 매핑
  @param iterable - list 또는 tuple
  @param func - 아규먼트 1개이고, 리턴값이 있는 함수
  @return 매핑된 값들을 갖는 리스트(list)
  """
  # result =[]
  # for x in iterable:

  #     result.append(func(x))
  result=[func(x) for x in iterable]

  return result
````
