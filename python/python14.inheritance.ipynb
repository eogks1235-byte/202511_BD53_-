# Inheritance(상속)

*   상속: 상위 클래스의 속성(변수)들과 기능(메서드)들을 하위 클래스에서 재사용(reuse)하는 것.
*   IS-A 관계까 성립하는 객체들을 상속 관계로 구현.
    *   학생은 사람이다. (Student IS A person.)
        * 사람 - 상위 클래스 (super class), 부모 클래스(parent class)
        * 학생 - 하위 클래스 (sub class), 자식 클래스(child class)
    *   상속: 상위 클래스는 하위 클래스에서 확장하는 것
*   HAS-A 관계가 성립하는 객체들은 일반적으로 상속으로 구현하지 않고, 속성을 사용해서 클래스를 구현.
    *   학생은 시험 점수를 갖는다.(Student HAS A score.)
        * 학생 클래스에서 속성(변수)들 중 하나로 Score 타입을 선언.
*   문법
    ```
    class SubClass(SuperClass):
          클래스 몸체(body)
    ```        

````python
#상위 클래스 선언
class Person:
  def __init__(self, name):
    self.name=name

  def say_hello(self):
    print(f'안녕하세요, 저는 {self.name}입니다.')
````

````python
#Person 타입의 객체를 생성
ssam= Person('오쌤')

#Person객체의 메소드를 출력
ssam.say_hello()
````

````python
#Person 클래스를 상속(확장)하는 Student클래스 선언
      #하위   상위
class Student(Person):
  def say_hello2(self):
    print('저는 학생입니다.')
````

````python
# Student 타입의 객체를 생성
gildong=Student('홍길동')
````

````python
#Student 타입 객체에서 메서드 호출
gildong.say_hello()
````

````python
gildong.say_hello2()
print(gildong.say_hello2())
#ssam.
````

# Method Override

*   상위 클래스가 가지고 있는 메서드를 하위 클래스에서 *같은 이름으로 재정의*하는 것.
    * 하위 클래스 타입의 객체에서는 하위 클래스에서 재정의한 메서드만 호출할 수 있음.
*   하위 클래스에서 override된 상위 클래스를 호출하기 위해서 `super().method_name(arg)` 형식으로 호출 가능함.(문법예제참고)      

````python
# #상위 클래스 선언
# class Person:
#   def __init__(self, name):
#     self.name=name

#   def say_hello(self):
#     print(f'안녕하세요, 저는 {self.name}입니다.')
````

````python
# 회사원은 사람이다. IS-A(사람-상위,,회사원-하위)
#         하위        상위
class BusinessPerson(Person):
  # __init__메서드 override
  def __init__(self,name,company): # name을 물려받음
    super().__init__(name) #상위 클래스 Person의 __init__메소드 호출
    self.company=company #상위 클래스에는 없고, 하위 클래스에만 있는 속성(변수)초기화
  # say_hello 메서드 override
  def say_hello(self):
    super().say_hello() #상위 클래스 Person의 say_hello메서드 호출
    print(f'저는{self.company}에 다닙니다')
````

````python
# BusinessPerson 타입의 객체를 생성
hgd=BusinessPerson('홍길동','아이티윌')
````

````python
hgd.say_hello()
````

## 상속& 메서드 override 연습

*   하이브리드 자동차는 자동차이다.
    *   자동차 -상위
    *   하이브리드 -하위
    

````python
#상위 클래스선언
class Car:
  def __init__(self,fuel):
    self.fuel=fuel
  def drive(self):
    print(f'자동차를 운전: 연료={self.fuel}L')
````

````python
# Car 타입의 객체를 생성
c1=Car(50)
````

````python
c1.drive()
````

````python
#Car클래스를 상속하는 HybridCar 클래스 선언
class HybridCar(Car):
#     상위 클래스 Car 의 __init__메서드를 override
  def __init__(self, fuel, battery):
    super().__init__(fuel)
    self.battery=battery
#      상위 클래스 Car의 drive메서드를 override
  def drive(self):
    print(f'하이브리드 자동차 운전하고 있습니다: 연료={self.fuel}L, 베터리={self.battery}%') #변수를 가져온다
````

````python
#HybridCar타입 객체 호출
c2=HybridCar(80,100)
````

````python
c2.drive()
````

# `isinstance(object, ClassName)` 함수

객체(object)가 어떤 클래스 타입의 인스턴스(생성된 객체)인 지 아닌 지(True/False)를 리턴하는 함수


````python
numbers=[1,2,3]
print(isinstance(numbers, list))  #> numbers가 list 타입의 인스턴스인가요? True
print(isinstance(numbers, dict))  #> numbers가 dict 타입의 인스턴스인가요? False
````

## `isinstance` 함수와 상속

````python
# c1 - Car 타입 객체(Car 생성자를 호출해서 생성한 객체)
print(isinstance(c1,Car))
print(isinstance(c1,HybridCar))
````

````python
 # c2 - HybridCar 타입 객체(HybridCar 생성자를 호출해서 생성한 객체)
print(isinstance(c2,Car))
print(isinstance(c2,HybridCar))  # 하이브리드가 상속받은 객체라서 True가 나오는듯= 그게맞는듯
````

## `isinstance` 함수활용

````python
# 강아지는 동물이다, 물고기능 동물이다. (동물 -> 상위클래스, 강아지/물고기 -> 하위클래스)
class Animal:
  def move(self):
    pass

class Dog(Animal):
  def move(self): #재 정의 override
    print('강아지가 총총총...')
class Fish(Animal):
  def move(self):
    print('물고기는 스윔스윔...')

class Tree:
  pass
````

````python
dog = Dog() #Dog타입의 객체를 생성
fish= Fish()#Fish타입의 객체를 생성
tree=Tree()#Tree타입의 객체를 생성

arr =[dog, fish, tree]
for x in arr:
  if isinstance(x,Animal): #상위타입검색
    x.move()
````

````python
class Shape:
  def __init__(self,x=0,y=0):
    self.point=Point(x,y)
#    self.point=x
 #   self.poitn=y
#self,
  def __str__(self):
    return f'도형 (x={self.point.x},y={self.point.y})'
  def move(self,dx,dy):
    self.point.move(dx,dy)

#  def area(self):
#도형은 점을 가지고 있다 Shape HAS-A point
class Shape:
  def __init__(self,x=0,y=0):
    """Point 타입 객체를 생성하기 위한 x좌표와 y좌표를 **아규먼트**(앞에사용한)로 전달받음"""
    self.point=Point(x,y)
    """Point 타입 객체를 아규먼트로 전달바듬"""
    #self.point=point #sh=Shape(pt1)포인트객체선택호출 (self,point)
  def __repr__(self):
    return f'Shape(point={self.point})'
  def move(self,dx=0,dy=0):#만약 무브함수를 안썻으면 sh.point.move(,)
    self.point.move(dx,dy)
    return self #move메서드를 호출한 객체(self)를 리턴함
  # def area(self):
  #   pass
  def area(self):
    raise NotImplementedError('넓이 계산 메서드를 구현해야함')
  def perimeter(self):
    raise NotImplementedError('둘레 길이 계산 메서드를 구현해야함')
````

````python
class Circle(Shape):
  def __init__(self, x=0, y=0, radius=0):
     super().__init__(x, y)
     self.radius = radius
  #  super().move(dy)
  def __str__(self):
    return f'원 (반지름={self.radius}, x={self.point.x},y={self.point.y})'

  def move(self,dx,dy):
    self.point.move(dx,dy)

  def area(self):
    return 3.14*(self.radius**2)
  def perimeter(self):
    return 3.14*(self.radius*2)
#T
#Shape를 상속하는 Circle클래스
class Circle(Shape):
  def __init__(self, x=0,y=0,radius=0):
    #super().class 객체를 생성(초기화)한다.
    super().__init__(x,y) #상위 클래스 객체 생성(초기화)--Shape클래스의 __init__
    #self.point=Point(x,y) 상속을 안받았을때
    self.radius=radius

  def move(self, dx,dy):
    self.point.move(dx,dy)
  #__str__ 상속이된다
  def __str__(self):#override
    return f'Circle(center={self.point},radius={self.radius})'
#                           self.point=Point(x,y)

  def area(self):
    return 3.14*(self.radius**2)
  def perimeter(self):
    return 3.14*(self.radius*2)

````

````python
class Rectangle(Shape):
  def __init__(self,x=0, y=0, width=0, height=0):
    # super().__init__(x)
    # super().__init__(y)
    super().__init__(x, y)
    self.width=width
    self.height=height

  def __str__(self):
    return f'직사각형(가로= {self.width}, 세로={self.height}, x={self.point.x},y={self.point.y})'
  def move(self,dx,dy):
    self.point.move(dx,dy)

  def area (self):
    return self.width*self.height
  def perimeter(self):
    return 2*(self.width+self.height)
#근데 왜 나는 값넣으면오류나냐?
#T

#Shape 클래스를 상속하는  Rectangle클래스
class Rectangle(Shape):
  def __init__(self,corner_x=0,corner_y=0,width=0,height=0):
    super().__init__(corner_x,corner_y)#--Shape클래스의 __init__
    self.width=width
    self.height=height
  def __repr__(self):
    return f'Rectangle(corner={self.point},width={self.width},height={self.height})'
  def area(self):
    return self.width*self.height
  def perimeter(self):
    return 2*(self.width+self.height)
````

````python
r1=Rectangle(width=2,height=2)
print(r1)
print(r1.area())
r1.move(1,2)
print(r1)
#T
rect=Rectangle(width=4,height=3)
print(rect)
print(f'직사각형 넓이={rect.area()}')
print(f'직사각형 둘레={rect.perimeter()}')
rect.move(-1,-1)
````
