# Ex. 3
아래의 두 2차원 배열 m1, m2에서 같은 인덱스의 원소들끼리 뺄셈(m1 - m2) 결과를 저장하는 리스트를 만들고 출력하세요.

m1 = [
    [1, 2, 3],
    [4, 5, 6]
]
m2 = [
    [1, 1, 1],
    [2, 2, 2]
]

```python
import numpy as np
```

```python
m1 = np.array([[1,2,3],
                [4,5,6]])
m2 = np.array([[1,1,1],
                [2,2,2]])
# m1 =(
#     [1, 2, 3],
#     [4, 5, 6]
# )
# m2 =(
#     [1, 1, 1],
#     [2, 2, 2]
# )
# m1 =(
#     [1, 2],
#     [4, 5]
# )
# m2 =(
#     [1, 1],
#     [2, 2]
# )
```

```python
print(m1,m2)  #같은 행끼리 계산
print(m1.sum())
print(m2.sum())
```

```python
# result=[
#     [r1-r2 for r1,r2 in zip(row1,row2)]
#     for row1, row2 in zip(m1, m2) # 1순위  크게 [] for x in zip 이 모양이다
# ]
# result
print(m1-m2)
```

```python
# 1차원 배열의 각 원소들끼리 뺄셈결과를 저장할 리스트
result=[]

for row1,row2 in zip(m1,m2):  # 1 크게 잡고
  row_minus=[]                # 3 작은거 넣기
  for x, y in zip(row1, row2): # 2 작은거 계산
    row_minus.append(x-y)
  result.append(row_minus)
result

#처럼 행렬 형태(2차원 리스트)를 유지하기 위해
#각 행을 계산하여 행 단위로 묶는 중간 리스트가 필요합니다.
```
