# while 반복문

```
[초기화 문장]
while 조건식:
    조건식을 만족하는 동안에 반복할 코드 블록
    [조건을 변경할 수 있는 문장]
```
[생략가능]

*   조건식을 만족하는 동안에 블록 내부를 반복해서 실행하게 됨.
*   조건식을 만족하지 못하면(조건식이 False가 되면) 반복을 멈춤.
*   `while 조건식` 다음에는 반드시 콜론(:)을 사용해야 됨
*   블록 안에서는 들여쓰기 크기를 일정하게 작성.


````python
# 1 ~ 5까지 순서대로 출력
print(1)
print(2)
print(3)
print(4)
print(5)
````

````python
# 1 ~ 5까지 순서대로 출력
n=1 #변수 초기화 - 변수를 선언하고, 최초로 값을 할당(저장)하는것
while n<=5:
  print(n*n-1)
  n += 1
````

````python
n=5
while n > 0: #n>=1
  print(n)
  n -= 1 # 들여쓰기 전부다 해줘야함
````

# `break`와 `continue`

*   `break`: 반복(iteration)을 종료. break가 포함된 가장 가까운 반복문만 빠져나감.


*   `continue`: 다음 반복(iteration)을 계속 수행. continue 아래의 문장들은 건너뛰고, 그 다음 반복을 수행.


````python
n=1
while n<6:
  if n==3:
    break
  print(n)
  n+=1# 3에서 if true라서 그만출력? ㅇㅇ
````

````python
n=0
while n<5:
  n+=1
  if n ==3:
    continue
  print(n)#3에서 print를 건너뜀
````
