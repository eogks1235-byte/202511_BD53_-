# Dictionary (dict {} )

*   사전(dictionary) 형식의 데이터 타입
*   키(key)와 값(value)의 쌍으로 저장하는 데이터 타입
    *   list, tuple: 인덱스를 기반으로 값들을 저장하는 데이터 타입
    *   dict: 키를 기반으로 값을 저장하는 데이터 타입
    *   dict타입에서 키의 역할은 list나 tuple의 인덱스의 역할과 비슷하다
        * dict의 저장된 1개의 값을 참조할때.
        * 키는 중복된 값을 허용하지 않는다


```python
#key값만 중복안되게끔
students={100: '홍길동', 101: '오쌤', 200: 'Scott', 201: '홍길동'}
print(type(students))
print(students)
```

```python
# 1번째 dict_name[key] -> value를 반환
students[100]
```

```python
# students[300]  error
#-> dict에 존재하지 않는 키를 사용하면 KeyError가 발생
```

# dict 매서드

```python
# dict_name.keys(): dict가 가지고 있는 키들의 리스트
students.keys()
```

```python
# dict_name.values(): dict가 가지고 있는 값들의 리스트
students.values()
```

```python
# dict_name.items(): (key, value) 튜플들의 리스트
students.items()
```

```python
# dict_name.get(key): key에 해당하는 값(value)를 리턴(반환)
# dict_name[key]과 비슷 get() 매서드는 존재하지 않는 키 일때 에러가 발생은 안하는데 출력도 없음
print(students.get(100))
print(students.get(400)) #none
```

`dict_name[key] = value`

*   key가 dict에 있을 때, key에  매핑된 값을 새로운 값으로 변경 10, 길동 > 10, 홍길
*   key가 dict에 없을 때, key=value 아이템을 dict에 추가


```python
print(students)
```

```python
#*   key가 dict에 있을 때, key에  매핑된 값을 새로운 값으로 변경
students[201]='김길동'
print(students)
```

```python
#*   key가 dict에 없을 때, key=value 아이템을 dict에 추가
students[300]='King' #키가 존재하지 않을때
print(students)
```

# dict와 for 반복문

```python
# for- in 반복문에서 dict를 사용하면 key들을 순서대로 반복(iteration)
for h in students:
  print(h, ':', students[h]) #*   key가 dict에 있을 때, key에  매핑된 값을 새로운 값으로 변경
              #students.get(h)=value
```

```python
for k, v in students.items():
  print(k,':',v)
```

```python
#둘다문자열  타입 달라도 ok
book ={'title' : '파이썬 예제', 'author' : '오쌤', 'price' : 10_000}
print(book)
#문자열은 따옴표필요
print(book['title'])
print(book.get('author'))
```

```python
for k, v in book.items():
  print(k, ':', v)
```

# 다양한 타입의 자료들을 저장하는 dict

```python
contact ={
    'id' : 1,
    'name' : '오쌤',
    'phone' : ['010-0000-0000', '02-0000-0000', '02-1234-5678'],
    'email' : {'company' : 'jake@itwill.com', 'personal' : 'jake@gmail.com'}
}
print(contact)
```

```python
contact['id'] #contact.get('id')
```

```python
#첫번째 전화번호 dict list index    list 문자열에서 index
print(contact['phone'])
print(contact['phone'][0])
```

```python
contact.get('phone')[0]
```

```python
#회사 이메일                          dict문자열 key
contact['email']['company']
```

```python
contact.get('email').get('company')
```

# Dictionary Comprehension

```python
emp_no = [1001, 1002, 1003, 1004] #사번
emp_name = ['Scott', 'Allen', 'King', '오쌤'] #사원이름

employees={} #(사번, 이름) 아이템(key, value)을 저장하기 위한 빈 dict

# for (k1,v1), (k2,v2) in zip(emp_no.items(), emp_name.items()):
#   print( v1, v2)
# employees
for k, v in zip(emp_no, emp_name):
  employees[k]=v
employees
#*   key가 dict에 없을 때, key=value 아이템을 dict에 추가
```

```python
employees2 = {no: name for no, name in zip(emp_no, emp_name)}
employees2
```
