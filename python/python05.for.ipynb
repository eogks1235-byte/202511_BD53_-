# for 반복문

```
for 변수 in iterable_type:
    반복해서 실행할 코드 블록
```

*   iterable_type : 반복(iteration)이 가능한 데이터 타입.
    * str, range, list, tuple, dict, ....
*   for 반복문은 iterable type에서 아이템들을 순서대로 하나씩 꺼내서 변수에 저장하고 코드블록 실행
*   for 반복문은 iterable type에서 더이상 꺼낼 아이템이 없을 때 종료됨


````python
for s in '안녕하세요' :
  print(s) # in 범위내 무언갈 순차적으로 쪼개서 하나씩 출력
````

# for 반복문과 `range()` 함수

````python
# range(start, end): start가 end보다 작은 경우, [start, end) 범위에 1씩 증가하는 정수들을 만듦.
for x in range(1, 6):          #6-1 = 5
  print(x)
````

````python
#range(end): 0부터 end까지 [0:end) 범위의 1씩 증가하는 정수를만듦
for x in range(6):          #6-0 = 6
  print(x)
````

````python
#range(start, end, step): step>0 start가 end보다 작고 step은 양의 정수인 경우,
#[start,end) 범위의 step 만큼씩 증가하는 정수들을 만듦.
for x in range(0,11,2):     # 짝
  print(x)
````

````python
for x in range(1,11,2):     # 홀
  print(x)
````

````python
# range(start, end, step): start가 end보다 크고 step이 음의 정수인경우,
# start >= x > end 인 범위의 step 만큼씩 감소하는 정수들을 만듦.
# for x in range(5,0,-1):
#   print(x)
for x in range(5,0,-1):
  print(x)
````

# break vs continue

* `break`: iteration을 멈춤.
* `continue`: 그다음 iteration을 수행.

````python
for x in range(5):
  if x==2:
    break
  print(x)
````

````python
for x in range(5):
  if x ==2:
    continue
  print(x)
````

````python
for y in range(2,10):
  print(f'==={y}단===')
  for x in range(1,10):
    if y+1==x:
      break   #if가 print보다 앞에있어서 증가하자마자 정지되는 바람에 +1을해버림
    print(f'{y} x {x} = {y*x}')
````

# `print()` 함수

*   탈출 문자열(escape string):
    * `'\n'`: 줄바꿈(new line)
    * `'\t'`: 탭(tap)
    * `'\''`: 작은 따옴표. 문자열 안에 인용부호를 포함하고 싶을 때 사용
* `print()` 함수의 파라미터(parameter)
    * `end`: print 함수가 아규먼트들을 출력한 후 마지막에 어떤 문자열을 추가할 것인 지를 설정. 기본값은 줄바꿈(`\n`)
    * `sep`: separator(구분자). print 함수가 아규먼트들을 출력할 때, 각각의 아규먼트들을 구분하기 위한 문자열을 설정. 기본값이 공백(`' '`) 1개



````python
print('I\'m a boy')
````

````python
print('안녕하세요')
print('월요일 점심은?')
````

````python
print('안녕하세오', end='::') # \n >> :: 줄바꿈이 디폴트값이다
print('월요일 점심은?')       # \n
print('hello~')
````

````python
print('안녕', 'hello')
````

````python
print('안녕', 'hello', sep=',')
````

````python
for x in range(1, 101):
  n_1 = x % 10 # 일의 자리 숫자
  n_10 = x // 10 # 십의 자리 숫자

  # 일의 자리 또는 십의 자리에 3, 6, 9 중 하나라도 포함되면 # 출력
  if (n_1 == 3 or n_1 == 6 or n_1 == 9) or \
     (n_10 == 3 or n_10 == 6 or n_10 == 9):

    print('#', end='\t')
  else:
    print(x, end='\t')

  # 10개마다 줄바꿈
  if x % 10 == 0:
    print()
````

````python
for x in range(1, 101):
    n_1 = x % 10  # 10으로 나눈 나머지 -> 1의 자릿수
    n_10 = x // 10  # 10으로 나눈 몫 -> 10의 자릿수
    b_1 = (n_1 == 3) or (n_1 == 6) or (n_1 == 9)  # 1의 자릿수가 3 또는 6 또는 9이면
    b_10 = (n_10 == 3) or (n_10 == 6) or (n_10 == 9)  # 10의 자릿수가 3 또는 6 또는 9이면

    if b_1 and b_10:
        print('##', end='\t')
    elif b_1 or b_10:
        print('#', end='\t')
    else:
        print(x, end='\t')

    if x % 10 == 0:
        print()  # 줄바꿈
````
