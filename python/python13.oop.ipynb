# 함수형 프로그래밍(Functional Programming)

```python
def create_student(id, name, kor, eng, math):
  student={
      'id': id,
      'name' : name,
      'korean':kor,
      'english':eng,
      'math':math
  }

  return student
```

```python
students=[
    create_student(1,'홍길동',100,90,95),
    create_student(2,'오쌤',90,50,80),
    create_student(3,'SCOTT',80,100,95)
]
```

```python
# # 아규먼트 student가 create_student()함수에 의해서 만들어진 dict일때 세과목의 총점을리턴
# def get_total(student):
#   return student['korean']+student['english']+student['math']
def get_total(student):
  return student['korean']+student['english']+student['math']
```

```python
# 아규먼트 student가 create_student()함수에 의해서 만들어진 dict일때 세과목의 평균을 리턴
def get_mean(student):
  return get_total(student)/3
```

# 객체 지향 프로그래밍(OOP: Object-Oriented Programming)

* **객체 (Object)**: 데이터(속성)와 기능을 가지고 있는 대상.
* **클래스 (class)**: 객체를 만들기 위한 설계도(코드).
    * 속성(attribute): 객체가 가져야하는 값(들), 변수(들).
    * 기능(method): 클래스 안에서 선언하는 함수(function)
      * 매서드를 선언할 때 항상 첫번째 파라미터는 `self`.
* **인스턴스(instance)**: 메모리에 생성된 객체.
* **생성자(constuctor)**: 클래스 이름과 동일한 특별한 함수. 객체를 생성해 주는 함수.
    * 파이썬에서는 생성자를 호출하면 클래스에서 선언된 `__init__()` 매소드가 자동으로 호출됨.

* 함수안에 기능이 포함되어있음    
* 붕어빵-객체
* 붕어빵틀-클래스 (코드)
 반죽 팥 >>속성
* 구워진붕어방 -인스턴스 ()

```python
#      클래스
class Student:
  # self시작 생성자함수-객체 초기화
                  #생성자__init__같은
  def __init__(self, id, name, korean, english, math):
    #인스턴스
    self.id =id
    self.name=name
    self.korean=korean
    self.english=english
    self.math=math

  # 메서드(method)
  # 학생 객체가 가지고 있는 세 과목의 총점을 리턴하는 메서드
  def get_total(self):
    return self.korean+ self.english + self.math

  # 학생 객체가 가지고 있는 세 과목의 평균을 리턴하는 메서드
  def get_mean(self):
    return self.get_total() / 3
```

```python
# Student 클래스의 생성자를 호출해서 객체를 생성   student1=인스턴스
student1=Student(1,'홍길동',99,85,96)
print(student1) #객체출력
print(student1.id) #객체속성출력
print(student1.name)
print(student1.korean)
print(student1.english)
print(student1.math)

print(student1.get_total()) #객체메서드출력
print(student1.get_mean())
```

```python
student2=Student(2,'오쌤',90, 80, 70)
print(student2)
print(student2.id)
print(student2.name)
print(student2.korean)
print(student2.english)
print(student2.math)
print(student2.get_total())
print(student2.get_mean())
```

## Ex.1

원(circle) 클래스

*   속성(attribute): 반지름(radius)
*   기능(method): 넓이 계산(area)3.14*r**2, 둘레 길이 계산(perimeter)2*r*3.14

```python
# class Circle:
#   def __init__(self, radius):
#     self.radius=radius

#   def area(self):
#     return (self.radius**2)*3.14

#   def perimeter(self):
#     return 2*self.radius*3.14
class Circle:
  def __init__(self,radius):
    self.radius=radius
  def area(self):
    return (self.radius**2)*3.14
  def perimeter(self):
    return 2*(self.radius)*3.14

```

```python
c1=Circle(1) # 생성자 호출 >> 객체 생성 ==c1=인스턴스
print(c1)
print(c1.radius) # 생성된 객체(인스턴스)의 속성(변수) 값을 읽어옴
print(c1.area()) # 생성된 객체(인스턴스)에서 메서드를 호출(기능을 호출) method는 괄호 필요
print(c1.perimeter())
```

```python
c2=Circle(10)   #self= c2 인스턴스 인스턴스 내에 정보를 이용
print(c2)
print(c2.radius)
print(c2.area())
print(c2.perimeter())
```

직사각형(Rectangle) 클래스

*   속성: width(가로길이), height(세로길이)
*   기능:
    * area: 직사각형 객체의 넓이 리턴 가로*세로
    * perimeter: 직사각형 객체의 둘레길이 리턴 2*(가로+세로)
    * resize: 직사각형 객체의 가로길이와 세로길이를 변경 no return ,새로운아규먼트

```python
# class Rectangle:
#   def __init__(self, width, height):
#     # 값 초기화(initializatiton)
#     self.width=width
#     self.height=height


#   def area(self):
#     return self.width*self.height
#   def perimeter(self):
#     return (self.width+self.height)*2
#   def resize(self,n_width,n_height): #파라미터
#     self.width=n_width
#     self.height=n_height

class Rectangle:
  def __init__(self, width,height):
    self.width=width
    self.height=height

  def area(self):
    return self.width*self.height
  def perimeter(self):
    return (self.width+self.height)*2
  def resize(self,n_width,n_height):
    self.width=n_width
    self.height=n_height
```

```python
# class Point:
#   """2차원 평면에서 한점의 좌표(x,y)를 표현하는 클래스"""
#   def __init__(self,x1,y1):
#     self.x1=x1
#     self.y1=y1

#   def move(self,dx,dy):
#     self.x1+=dx
#     self.y1+=dy

#   def distance(self,x2,y2):

#     # self.x2=x2
#     # self.y2=y2

#     # return ((self.x1-self.x2)**2+(self.y1-self.y2)**2)**0.5
#     return ((self.x1-x2)**2+(self.y1-y2)**2)**0.5
class Point:
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y
  def move(self,dx,dy):
    self.x=self.x+dx
    self.y=self.y+dy
  def distance(self,other):
    return ((self.x-other.x)**2+(self.y-other.y)**2)**0.5
```

```python
class Point:
  """2차원 평면에서 한점의 좌표(x,y)를 표현하는 클래스"""
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y

  def move(self,dx,dy):
    """점의 위치를 x축의 방향으로 dx만큼, y축의 방향으로 dy만큼 이동"""
    self.x+=dx
    self.y+=dy

#   def distance(self,x2,y2):
#     """ㅇ"""
#     self.x2=x2
#     self.y2=y2

#     return ((self.x1-self.x2)**2+(self.y1-self.y2)**2)**0.5
# # def distance(self,point):
#   self.=point

  # def distance(self,x,y):
  #   return ((self.x1-x)**2+(self.y1-y)**2)**0.5
  def distance(self,other):
    """현재 점(self)과 다른 점(other)사이 거리를 리턴"""
    """
    @param other - Point 타입객체
    @return self 와 other 사이의 거리
    """
    #return 과 저장값 동시에 사용가능
    return ((self.x-other.x)**2+(self.y-other.y)**2)**0.5
```

```python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age

  def __str__(self):
    return f'Person(name={self.name}, age={self.age})'

  def __repr__(self):
    return f'이름={self.name}, 나이={self.age}'

```

# 클래스 작성 연습

*   Score 클래스
    * 속성: 국어(korean), 영어(english), 수학(math)
        *   객체를 초기화할 때 기본값은 모두 0.
    * 메서드:
        *   총점 리턴(get_total)
        *   평균 리턴(get_mean)
        *   `__str__()`:`Score(korean=100, english=100, math=100)` 형식의 문자열을 리턴
        *   `__repr__()`:`점수(국어=0, 영어=0, 수학=0)` 형식의 문자열을 리턴

*   Student 클래스
    * 속성: 번호(id), 이름(name), 점수(score: Score클래스 타입을 저장하는 변수)
    * 메서드
        *   `__str__()`
        *   `__repr__()`

```python
class Score:
  def __init__(self, korean=0, english=0,math=0):
    self.korean=korean
    self.english=english
    self.math=math
  def get_total(self):
    return self.korean+self.english+self.math
  def get_mean(self):
    return self.get_total()/3
  def __str__(self):
    return f'Score(korean={self.korean}, english={self.english}, math={self.math})'
  def __repr__(self):
    return f'점수(국어={self.korean}, 영어={self.english}, 수학={self.math})'
```

```python
s1=Score(10,10,10)
print(s1)
print(s1.get_total())
print(s1.get_mean())
s1
```
