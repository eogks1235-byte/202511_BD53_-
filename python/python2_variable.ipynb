# Python Data Type(자료 타입)


* 숫자 타입
  *   정수(int) : integer
  *   실수(float) : floating-point number
* 문자열 타입(str) : string
* 논리 타입(bool) : True/False



## 숫자 타입

```python
# 변수 선언, 값 할당: 변수이름 = 값
n = 123
print(type(n))  #type(변수이름): 변수의 타입을 반환하는 함수
print('n=',n)
```

```python
n=3.141592
print(type(n))
print('n=', n)
```

## 문자열 타입

파이썬의 문자열은 작은따옴표('')와 큰따옴표("")를 모두 사용할 수 있다

```python
name = '오쌤'
print(type(name))
print('name =', name)
```

## 명시적인 타입 변환

*   `int(arg)`: 아규먼트(arg)로 전달된 값을 정수로 변환해서 반환하는 함수.
    * 문자열 >  정수
    * 실수   >  정수
*   `float(arg)`: 아규먼트로 전달될 값을 실수로 변환해서 반환하는 함수.    
    * 문자역 >  실수
    * 정수   >  실수
*   `str(arg)`: 아규먼트로 전달된 값을 문자열로 변환해서 반환하는 함수
    * 정수, 실수, ... > 문자열
        

```python
x= input('정수를 입력 >>> ')
y= input('정수를 입력 >>> ')
print('변수 x의 타입 :', type(x))
print('변수 y의 타입 :', type(y))
x=int(x) #x의 값을 int타입으로 변환하는 과정
y=int(y) #y의 값을 int타입으로 변환하는 과정
print('변수 x의 타입 :', type(x))
print('변수 y의 타입 :', type(y))

print(f'{x}- {y} = {x-y}')
```

# 연산자(Operator)

## 할당(assignment) 연산자

*   `=`: 연산자 오른쪽의 값을 왼쪽에 변수에 저장(할당)

## 산술(artismetic)연산자

*    종류: `+, -, *, /, **, //, %`
*    산술 연산자는 일반적으로 숫자 타입(int, float)에서만 사용
*    일부 산술 연산자는 문자열과 같은 숫자 이외의 타입에서도 사용 가능

```python
x, y = 10, 3
print(f'{x} + {y} = {x+y}')
print(f'{x} - {y} = {x-y}')
print(f'{x} x {y} = {x*y}')
print(f'{x} / {y} = {x/y}')
print(f'{x} 의 {y} 거듭제곱 = {x**y}')
print(f'{x} 를 {y}로 나눈 몫 = {x//y}')
print(f'{x} 를 {y}로 나눈 나머지 = {x%y}')
```

문자열에서 사용할 수 있는 산술 연산자:

*   `문자열 + 문자열` : concatenate. 문자열 이어 붙이기
*   `문자열 + 정수`, `정수 + 문자열` : replicate. 문자열을 정수 만큼 복사


```python
print('Hello' + 'Python')
print('Hello' * 3)
print(3 * 'Hello')
```

```python
x, y ='1', '2'
print(x+y)  # + 연산자: 문자열 이어붙이기 연산자
```

## 복합 할당 연산자

*     종류: `+=, -=, *=, /=, ...`
*     산술 연산자와 할당 연산자를 합쳐서 만들어진 연산자.

```python
salary= 100
print(salary)

# salary = salary + salary * 0.1
salary += salary * 0.1  # salary *= 1.1
print(salary)
```

## 비교 연산자, 논리 연산자

*   비교 연산자: `==, !=, >, >=, <, <=`
*   논리 연산자: `and, or, not`

```python
#비교 논리는 t/f로 나타남
#
```

```python
x, y = 100, 200
print(x == y)
print(x != y)
print(x > y)
print(x < y)
```

```python
print((x > 50) and (x < 150))
print(x)
```

```python
x= None  #SQL에서의 null이다
print(x==None)  #파이썬은 is 연산자가 따로 없다
print(x!=None)
```
