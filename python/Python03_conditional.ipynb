# 조건문(conditional statement)


## else 없는 조건문

```
if 조건식:
    조건식을 만족할 때(조건식이 참일때) 실행할 코드 블록(문장들)
```

*   `if 조건식` 뒤에는 반드시 콜론(:)을 써야 됨
*   if 블록은 반드시 들여쓰기(indentation)을 해야 됨
*   들여쓰기 크기는 일정해야 함.

````python
age = int(input('나이 입력 >>> '))
if age >= 0:
  print('hi man...')  #마지막줄 들여쓰기안하면 무조건출력

  print(f'your age is {age}...')#출력하고 싶은 값이 문자열과 같이있는경우 f {}이방법이용
````

## else 있는 조건문

```
if 조건식:
    조건식을 만족할 때 실행할 코드 블록
esle:
    조건식을 만족하지 못할 때 실행할 코드 블록    
```

````python
age = int(input('나이 입력 >>> '))
if age >=0:
    print(f'당신은 {age}살...')
else:
    print(f'{age}살 뭐라카노!!')
````

Falsy Values: 파이썬에서 false로 취급하는 값들.

*   숫자(int, float): 0
*   문자역(str): '' (빈 문자열, empty string)
*   빈 리스트(empty list)
*   None: 변수에 값이 할당되어 있지 않은 경우 (값이 없는 경우)를 표현하는 파이썬 예약어(키워드)


````python
number = int(input('정수를 입력 >>> '))
if number % 2:    # number를 2로 나눈 나머지가 있으면  0은 false(후) 1은 true(선) 라는데 그게 어떻게들어가게 되는거냐 >> 파이썬특징이다
  print('홀수')   # 설명이있었다
else:  # 그 외
  print('짝수')
````

## chained conditional statement

```
if 조건식1:
    (조건식1)이 참일 때 실행할 코드 블록
elif 조건식2:
    (조건식2)가 참일 때 실행할 코드 블록
.3
.4
.5
else:
    위의 모든 조건들이 거짓일 때 실행할 코드 블록
else생략가능        
```

*   `elif` 블록은 1개 이상이 있을 수 있다.
*   `else` 블록은 생략 가능
*   모든 블록 안에서 들여쓰기 크기는 같은 크기로 작성


* `else`는 조건없네

````python
# 숫자(float)를 입력 받아서 양수/0/음수 인지를 출력
x=float(input('숫자를 입력 >> ')) #T >> number

if x > 0:
  print('양수입니다')
elif x == 0:
  print('0입니다')
else:
  print('음수입니다')
````

*    세 과목 시험 점수를 입력받고 저장.
*    세 과목의 총점 계산 출력
*    세 과목의 평균 계산 출력
*    세 과목의 평균이 90점 이상이면 A, 80점 이상이면 B, 70점 이상이면 C, 그렇지 않으면 F라고 출력.

````python
score1= int(input('점수1을 입력 >> '))
score2= int(input('점수2을 입력 >> '))
score3= int(input('점수3을 입력 >> '))
#total=score1+score2+score3
#mean= total/3;
print(f'{score1}+{score2}+{score3} sum=' ,score1+score2+score3)
print(f'{score1}+{score2}+{score3} avg=' ,(score1+score2+score3)/3)

if (score1+score2+score3)/3>=90:
  print('A')
elif (score1+score2+score3)/3>=80:
  print('B')
elif (score1+score2+score3)/3>=70:
  print('C')
else:
  print('F')
````

````python
score1 = int(input('점수1을 입력 >> '))
score2 = int(input('점수2를 입력 >> '))
score3 = int(input('점수3을 입력 >> '))

total = score1 + score2 + score3 # 정수 3개 받아서 total에 저장
print('총점 : ', total)   # total의 출력물

mean = total / 3
print('평균 : ', mean)

if mean >=90:
  print('A')
elif mean >=80:
  print('B')
elif mean >=70:
  print('C')
else:
  print('F')
````

## 중첩 조건문

*   if, elif, else 블록 안에서 또다른 조건문이 포함된 경우.
*   들여쓰기를 특히 주의!!!
*   들여쓰기 방식에 따라서 전혀 다른 로직이 될 수도 있다
*   중첩 조건문은 논리 연산자(and, or)를 사용하는 조건문 또는 elif 블록으로  바꿀 수도 있다.

````python
number = float(input('숫자를 입력 >>> '))
if number >0:
  print('양수')
else:
  if number < 0:
    print('음수')
  else:
    print('0')
````

#match 문

```
match 변수(또는 식):
    case 값1:  #if
        변수(또는 식)이 값1과 같은 경우 실행할 코드 블록
    case 값2:  #elif
        변수(또는 식)이 값1과 같은 경우 실행할 코드 블록
    ...      
```

````python
user = int(input('1, 2, 3 선택 >>'))
match user:
  case 1:
    print('가위')
  case 2:
    print('바위')
  case 3:
    print('보')
````

````python
#0(일요일), 1(월요일), ..., 6(토요일)
day= int(input('0~6 선택 >> '))
match day:
  case 1 | 2 | 3 | 4 | 5 :
    print('학원에서 수업')
  case 0 | 6:
    print('집에서 복습')
````
