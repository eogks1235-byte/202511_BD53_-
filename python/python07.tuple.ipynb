# tuple

*   tuple(튜플): 저장된 아이템들을 변경할 수 없는 리스트.
*   인덱스 기반의 데이터 타입
    *   indexing: `tuple_name[index]`. 특정 인덱스 위치의 아이템을 읽어옴
    *   slicing: `tuple_name[start:end]`. 특정 인덱스 범위의 아이템들로 이루어진 새로운 튜플을 만듦
*   append, extend, insert, pop, remove, ... 원소를 추가/삭제/변경하는 기능들은 제공되지 않음
    

```python
numbers= (1, 2, 3, 1, 1, 2)
print(type(numbers))
print(numbers)
```

```python
#tuple|list.count(value):튜플 또는 리스트에서 value가 몇개 있는지 반환
numbers.count(11) #1의 개수
```

```python
#len(tuple|list): 튜플 또는 리스트의 원소개수를 반환
len(numbers)  # | 또는
```

```python
numbers.index(1) # index(값) 값이 어디에 위치하고 있는지알려줌
#tuple|list.index(value):튜플 또는 리스트에서 value가 처음 나오는 인덱스를 반환
```

#  tuple 과 for 반복문

```python
numbers =(1, 10, -2, -20, 30)
for x in numbers:
  print(x)
```

```python
for index,value in enumerate(numbers):
  print(index, ':', value)
```

```python
numbers1=(1, 2, 3)
numbers2=(10, 20, 30)

for x, y in zip(numbers1,numbers2):
  print(x,y)
```

## 2차원 튜플과 for 반복문

```python
#작은 행끼리의 덧셈부터 시작
for x in matrix:
  print(x)
```

```python
for row in matrix:
  for item in row:
    print(item,end='\t')
  print() #줄바꿈
```

```python
qwe=[1,2,3],[4,5,6]
asd=[5,6,7],[8,9,0]
print(qwe)
print(asd)
```

```python
#1.최종계산식 2. a b b c
res=[
    [r1-r2 for r1,r2 in zip(row1,row2)]
    for row1,row2 in zip(qwe,asd)
]
res
```

```python
cols1 = list(zip(*qwe))  # m1의 열들을 모은 튜플 리스트
cols2 = list(zip(*asd))   #zip안에 *이있는 이유 : 열끼리묶어줌

result = [[c1 + c2 for c1, c2 in zip(col1, col2)]
          for col1, col2 in zip(cols1, cols2)]
result
```

```python
qwe=[1,2,3],[4,5,6],[5,6,7]
asd=[5,6,7],[8,9,0],[6,7,8]
zxc=[5,6,7],[6,7,8],[5,6,7]
print(qwe)
print(asd)
print(zxc)
```

```python
res=[
    [r1-r2-r3 for r1,r2,r3 in zip(column1,column2,column3)]
    for column1,column2,column3 in zip(qwe,asd,zxc)
]
res
```
