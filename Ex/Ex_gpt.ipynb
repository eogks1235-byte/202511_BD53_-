# 업다운 게임

*컴퓨터가 1~100 사이 랜덤 숫자 하나 생성

*사용자가 숫자를 맞출 때까지 입력

*입력값보다 크면 "UP", 작으면 "DOWN" 출력

*맞추면 시도 횟수 알려주기

```python
import random
```

```python
answer = random.randrange(1,101)
count=0

print('1~100까지 임의숫자 맞추기')


while True:
  guess = int(input('숫자 입력>> '))
  count+=1

  if guess == answer:
    print(f'정답! {count}만에 맞춤')
    break
  elif guess > answer:
    print('낮아')
  else:
    print('높아')
```

```python
# #문제 5 — 문자열에서 숫자만 추출하기

# 문자열:

# "p3y7t1h0o9n"


# 이 문자열에서 숫자만 골라서 하나의 정수로 만들어라.

# 예시:

# 37109


# 힌트:

# 문자열 슬라이싱,

# for 반복문,

# 문자.isdigit() 활용 가능
```

```python
ex_str= input('문자열을 입력하세요 >>')
#숫자추출
number=''
for x in ex_str:
  if x.isdigit(): #숫자 인지 확인
    number +=x
#문자열 > 정수로
result = int(number)
print(result)
```

```python

jejang=[]
bada=input('5개 아무거나 ㄱ').split() # 숫자간 띄어쓰기 필요

for x in bada:
  jejang.append(int(x))
print(jejang)
jemax=jejang[0]
jemin=jejang[0]
for y in jejang:
  if y > jemax:
    jemax=y
  if y < jemin:
    jemin=y
print(jemax, jemin)

print(jejang[::-1])
```

```python
# 학생 이름과 점수 튜플을 만들어 출력

# 튜플 안에서 점수가 80점 이상인 학생 이름만 출력
```

```python
hakna=('앨런', '스콧', '킹',' 오쌤')
hakdot=(80, 40, 25, 95)

for x,y in zip(hakna,hakdot):
  if y>=80:
    print(x)
```

```python
# 학생 이름과 점수를 딕셔너리로 저장

# 점수 합계와 평균 계산

# 특정 학생 점수 조회
```

```python
students = {
    '앨런': 80,
    '스콧': 40,
    '킹': 25,
    '오쌤': 95
}

hapdot=sum(students.values())
pyeong=hapdot/len(students)

print(hapdot)
print(pyeong)

who = input("점수를 조회할 학생 이름 입력 >> ")

whoscore=students.get(who)
if whoscore is not None:
  print(f'그 사람의 이름은 = {who}')
else:
  print('아무도 없다')
```
