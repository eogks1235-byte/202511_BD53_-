```python
import random
wallet= int(input('지갑안에 금액'))

menu=[15000, 8000, 6000]
dinner=random.choice(menu)

if wallet >= dinner:
  wallet-= dinner
  print(f'저녁식대 {dinner}원')
  if wallet>=6000:
    print(f'남은돈이 {wallet} 점심도 든든')
  else:
    print(f'점심 먹기 힘들어 {wallet}')
else:
  print(f'저녁식대 {dinner}원')
  if wallet >=6000:
    print(f'저녁은 못먹고 {wallet}으로 내일점심')
  else:
    print(f'저녁은 못먹고 {wallet}으로 내일점심도 힘음')
```

```python
#최대공약수
a= int(input('첫번째 숫자 :'))
b= int(input('두번째 숫자 :'))

for x in range(max(a,b),1, -1):# 최댓값 1까지 역순
  if a% x ==0 and b%x==0:
    print(f'최대공약수{x}')
    break
  else:
    continue
```

```python
# index를 이용해서 출력해보기
# for, if,elif,else
# vip이상은 발랫
# platinum은 우대주차
# family는 지하 12층
number=[['vvip', 50_000_000],['vip', 30_000_000],['platinum', 20_000_000],['family', 10_000_000]]

for x in number:
  if x[1] >= 30_000_000:
    print(f'{x}등급')
    print('발랫')
  elif x[1] >= 20_000_000:
    print(f'{x}등급')
    print('우대주차')
  elif x[1]>= 10_000_000:
    print(f'{x}등급')
    print('지하 4층 주차')
  else:
    print('지하12층')
```

```python
#list, dict, while
#세뱃돈구하기
#새뱃돈 금액 ㅡ평균
#할아 할머 아 엄 10 5 3 1 dict
#len(), zip()
#while
family=['할아버지','할머니','아빠','엄마']
money=[100_000,50_000,30_000,10_000]

bdict= dict(zip(family, money))
bdict
```

```python
cnt=len(family)-1
bdict=0

while cnt != -1:
  print(f'{family[cnt]}께서 {money[cnt]}')
  bdict+=money[cnt]
  cnt-=1

print(f"올해 새뱃돈{bdict} , 평균{bdict/len(family)}")
```

```python
#1~45 중복없이 6개 추출 한 리스트
#random
#무한루프 break 탈출
#in, not in

import random
lotto=[]
#loto=[random.randrange(1,46)for x in lotto]
while True:
  number = random.randint(1,46)
  if number not in lotto:
    lotto.append(number)
  lotto.sort()
  if len(lotto)==6:
    break
lotto
```

```python
import random
lotto=[]
#loto=[random.randrange(1,46)for x in lotto]
for _ in range(5):
    one=[]
    while True:
      number = random.randint(1,45)
      if number not in one:
        one.append(number)
      one.sort()
      if len(one)==6:
        break
    lotto.append(one)
lotto
```
