```python
class Circle:
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    return (self.radius**2)*3.14

  def perimeter(self):
    return 2*(self.radius)*3.14
```

```python
c1=Circle(1)
print(c1)
print(c1.radius)
print(c1.area())
print(c1.perimeter())
```

```python
class Point:
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y
  def edong(self,dx,dy):
    self.x+=dx
    self.y+=dy
  def geri(self,other):
    return ((self.x-other.x)**2+(self.y-other.y)**2)**0.5
```

```python
# class Score:
#   def __init__(self,korean=0,math=0,science=0):
#     self.korean=korean
#     self.math=math
#     self.science=science

#   def __str__(self):
#     return f'Score(korean={self.korean},math={self.math},science={self.science})'

#   def __repr__(self):
#     return f'점수(국어={self.korean}, 수학={self.math}, 과학{self.science})'

#   def get_total(self):
#     return self.korean+self.math+self.science

#   def get_mean(self):
#     return self.get_total()/3
class Score:
  def __init__(self, korean=0,math=0,science=0):
    self.korean=korean
    self.math=math
    self.science=science
  def __str__(self):
    return (f'Score(korean={self.korean},math={self.math},science={self.science})')
  def __repr__(self):
    return (f'점수(국어={self.korean},수학={self.math},과학={self.science})')
  def get_total(self):
    return self.korean+self.math+self.science
  def get_mean(self):
    return self.get_total()/3
```

```python
# class Student:
#   def __init__(self,id,name,score):
#     self.id=id
#     self.name=name
#     self.score=score
#   def __str__(self):
#     return f'Student(id={self.id},name={self.name},score={self.score})'
#   def __repr__(self):
#     return self.__str__()
class Student:
  def __init__(self,id,name,score):
    self.id=id
    self.name=name
    self.score=score
  def __str__(self):
    return (f'Student(id={self.id},name={self.name},score={self.score})')
  def __repr__(self):
    return self.__str__()
```

```python
# score1=Score()
# print(score1)
# score1
score1=Score()
print(score1)
score1
```
