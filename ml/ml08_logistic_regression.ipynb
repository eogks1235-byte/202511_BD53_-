# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report
```

# 데이터 셋

```python
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

물고기의 모든 특성을 사용한 이진 분류(Bream vs Smelt)

```python
df =fish[fish.Species.isin(['Bream','Smelt'])]
```

```python
df.head()
```

```python
x=df[df.columns[1:]].to_numpy() # 특성배열
y=df['Species'].to_numpy() # 타겟배열
```

## 훈련/테스트 나누기

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=42,stratify=y)
```

# Logistic Regression

## 모델 훈련

```python
# 표준화 + Logistic 분류 -> 파이프라인
pipe=Pipeline(steps=[('scaler',StandardScaler()),('clf',LogisticRegression(random_state=42))])
```

```python
# ML 모델 훈련
pipe.fit(x_train,y_train)
```

```python
pipe['scaler'].mean_ # 훈련 셋 물고기 5개 특성 평균
```

```python
pipe['scaler'].var_ # 훈련 셋 물고기 5개 특성 분산
```

```python
pipe['clf'].coef_ # 물고기 5개 특성 변수들 앞에 곱해주는 계수들(기울기)
```

```python
pipe['clf'].intercept_ # 선형회귀식의 절편
```

## 모델 평가

```python
# 훈련 셋 예측값
train_pred=pipe.predict(x_train)
train_pred
```

```python
test_pred=pipe.predict(x_test)
test_pred
```

```python
y_train # 훈련 셋 실젯값
```

```python
y_test
```

```python
train_pred_prob=pipe.predict_proba(x_train)
train_pred_prob[:7] # 도합 100%
# (n_samples, 2) shape의 2차원 배열 - [도미(Bream)일 확률, 빙어(Smelt)일 확률]을 아이템으로 갖는 2차원 배열.
```

```python
test_pred_prob=pipe.predict_proba(x_test)
test_pred_prob[:7]
```

```python
# 훈련 셋 혼동 행렬
cm_train=confusion_matrix(y_true=y_train,y_pred=train_pred)
cm_train
```

```python
cm_test=confusion_matrix(y_true=y_test,y_pred=test_pred)
cm_test
```

```python
labels=['Bream','Smelt']
sns.heatmap(cm_train,cmap='Reds',annot=True,xticklabels=labels,yticklabels=labels)
plt.show()
```

```python
sns.heatmap(cm_test,cmap='Blues',annot=True,xticklabels=labels,yticklabels=labels)
plt.show()
```

# Sigmoid(Logistic) 함수

$$
\phi(z) = \dfrac{e^z}{e^z + 1} = \dfrac{1}{1 + e^{-z}}
$$

```python
z = np.arange(-5, 5, 0.0001)
sigmoid = 1/ (1+np.exp(-z))


plt.figure(figsize=(8,4))
plt.plot(z, sigmoid)
plt.grid()
plt.xlabel('z')
plt.title('sigmoid')
plt.ylabel(r'$\phi(z)$') # 경고없이 그리스문자 사용은 문장앞 r적는다
plt.show()
```

## Logistic Regression

**Logistic Regression**

*   선형 회귀
$$
z_i = w_0 + x_{i1} \cdot w_1 + x_{i2} \cdot w_2 + \cdots = w_0 + \sum_j x_{ij} \cdot w_j
$$

*   선형 회귀 공식으로 계산된 결과를 logistic 함수의 argument로 전달
$$
p_i = \phi(z_i) = \dfrac{1}{1 + e^{-z}}
    = \dfrac{1}{1 + exp(-(w_0 + \sum_j x_{ij} \cdot w_j))}
$$

*   Sigmoid(Logistic) 함수의 리턴값은 항상 0 ~ 1이므로, Logistic Regression은 logistic 함수의 리턴값을 양성이 될 확률로 해석.(분류문제)
*   Logistic Regression에서 훈련(학습)이란 훈련 데이터로부터 아래의 비용 함수를 최소화하는 파라미터들($w_0$, $w_1$, ...)을 찾는 과정.

*   다중 클래스 분류에스는 softmax 함수를 클래스 $i$가 될 확률로 정의. 0<(i=컬럼?,j=컬럼들의합)<1
$$
\mathrm{softmax}{\mathbf(z)}_i = \dfrac{e^{z_i}}{\sum_{j=1}^{k} e^{z_j}}, \quad i = 1, \dots, k
$$


**비용(손실) 함수(cost/loss function)**

*   이진 분류에서는 손실 함수 log loss가 최소가 되는 계수들을 찾는 것이 목적.
*   이진 분류(p, 1-p = 1)
*   예측 실패에 대한 값
*   y 값을 p의확률로
*   Logistic loss function(**log loss**)

>>   $ L = -[y \log{p} + (1-y) \log{(1-p)}] $

>>   $ J(W) = - \dfrac{1}{N} \sum_{i=1}^{N} [y^{(i)} \log({p}^{(i)}) + (1-y^{(i)}) \log(1 - {p}^{(i)})] $

>>   N: the number of samples

*   다중 클래스 분류에서는 비용 함수 크로스 엔트로피가 최소가 되는 계수들을 찾는 것이 목적.
*   __Cross entropy__ cost function

>>   $ J(W) = - \dfrac{1}{N} \sum_{i=1}^{N} \sum_{k=1}^{m} y_{k}^{(i)} \log({p}_{k}^{(i)}) $

>>   N: the number of samples

>>   m: the number of classes

```python
# 상용 로그(common log, 밑수가 10인 로그) 시각화
p = np.arange(0.0001, 2.0, 0.0001)
y= np.log10(p)
plt.plot(p,y,label=r'$log_{10}(p)$')
plt.plot(p,-y,label=r'$-log_{10}(p)$')

plt.xlabel('p')
plt.grid()
plt.legend()
plt.show()
```

# 다중 클래스 분류

물고기 종류(7개 클래스: Perch, ... )를 여러가지 특성(5개 특성: Weight, ...)들로 예측

```python
fish.Species.value_counts()
```

```python
x= fish[fish.columns[1:]].to_numpy()
y=fish.Species.to_numpy()
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape
```

```python
pd.Series(y_train).value_counts()
```

```python
pd.Series(y_train).value_counts(normalize=True)
```

```python
pd.Series(y_test).value_counts(normalize=True)
```

```python
# Pipeline 객체 (StandardScaler, LogisticRegression) 생성
pipe = Pipeline(steps=[('scaler',StandardScaler()),('clf',LogisticRegression(random_state=42))])
```

```python
pipe.fit(x_train,y_train)
```

```python
# 훈련 셋 예측값 찾기
trainpred=pipe.predict(x_train)
trainpred[:6]
```

```python
train_prob=pipe.predict_proba(x_train)
train_prob[:6]
```

```python
y_train[:6]
```

```python
cm_train=confusion_matrix(y_train,trainpred)
cm_train
```

```python
sns.heatmap(data=cm_train,cmap='Blues',annot=True)
plt.show()
```

```python
pipe['scaler'].mean_
```

```python
pipe['scaler'].var_
```

```python
pipe['clf'].coef_ #.shape(7,5) =(n_classes, n_features)
```

```python
pipe['clf'].intercept_ #.shape(7,) = (n_classes,)
```

```python
# 훈련 셋 정확도
pipe.score(x_train,y_train)
```

```python
test_pred=pipe.predict(x_test)
```

```python
cm_test=confusion_matrix(y_test,test_pred)
```

```python
sns.heatmap(data=cm_test,cmap='Reds',annot=True)
plt.show()
```
