## ravel vs flatten

* ravel
  * n차원 배열을 **1차원 배열로 변환한 "뷰(view)"를 리턴.**
  * **새로운 1차원 배열을 만드는 것이 아니라**, 원본 배열의 모양을 1차원처럼 보여줄 뿐.
  * 리턴된 1차원 배열에서 **원소를 변경하면, 원본 배열도 변경됨!**
* flatten
  * n차원 배열을 **1차원 배열로 변환한 "복사본"을 리턴.**
  * **원본 배열과는 별개의** 새로운 1차원 배열을 메모리에 만들어서 리턴.  
  * 리턴된 1차원 배열의 원소를 **변경하는 것은 원본 배열에 영향을 주지 않음!**

```python
arr_2d=np.arange(12).reshape((3,4))
print(arr_2d)
```

```python
flattened=arr_2d.flatten() # 2d >> 1d
```

```python
flattened
```

```python
# 복사본
flattened[0]=123
print(flattened)
```

```python
print(arr_2d) # 원본 2d array는 변경되지 않았다.
```

```python
raveled=arr_2d.ravel() # 2d >> 1d
```

```python
print(raveled)
```

```python
raveled[0]=123
print(raveled)
```

```python
print(arr_2d) # 원본 2d array는 변경 되었다.
```

# 배열 합치기(concatenate)

*   `np.concatenate([array1,array2,...],axis=0)`(axis기본값=0, or 1)
    :특정 축(axis) 방향으로 배열 이어붙이기.

*   `np.r_[array1,array2]`: 행들을 이어붙이기. axis=0 방향으로 이어붙이기.(concatenate기준)
*   `np.c_[array1,array2]`: 컬럼들을 이어붙이기. axis=1 방향으로 이어붙이기.(concatenate기준)

```python
# (2,3) shape의 2d array
arr1 = np.arange(6).reshape(2,3)
print(arr1)
```

```python
# (1,3) shape의 2d array
arr2 = np.array([[10,20,30]])
print(arr2)
```

```python
# (2, 1) shape의 2d array
arr3 = np.array([[11],[12]])
print(arr3)
```

```python
result = np.concatenate([arr1,arr2]) #axis=0 은 기본이다
print(result)
```

```python
result = np.concatenate([arr1,arr3],axis=1) #axis=0 은 기본이다 2차원이상에서 다가능
print(result)
```

```python
np.r_[arr1,arr2] # 2차원만 가능하다
```

```python
np.c_[arr1,arr3]
```

# Broadcasting

*   numpy.ndarray의 연산(+, -, *, /, ==, !=, >, <, ...)은 같은 인덱스(위치)의 원소(element-wise)들끼리 연산이 수행됨.
*   서로 모양(shape)이 다른 배열끼리도 element-wise 연산이 가능한 경우가 있음 >> broadcast
*   broadcasting이 가능한 경우
    *   누락되었거나 길이가 1인 차원에 대해서 broadcasting이 수행.
    *   이어지는 차원의 축의 길이가 같은 경우에 broadcasting이 가능.

## broadcast 활용

*   표준화(standardization): 평균이 0이 되고 표준편차가 1이 되도록 값들을 변환.
>   $$z = \dfrac{x - \mu}{\sigma}$$ (mu=mean)(sigma=std)
*   정규화(normalization): 최솟값이 0이 되고 최댓값이 1이 되도록 값들을 변환.
>   $$x_{norm} = \dfrac{x - x_{min}}{x_{max} - x_{min}}$$

### 1차원 배열 표준화

```python
x = np.arange(1,6)
print(x)
```

```python
mu= x.mean() # 평균
print(mu)
```

```python
sigma=x.std() # 표준편차(standard deviation)
print(sigma)
```

```python
x_scaled = (x - mu) / sigma
print(x_scaled)
```

```python
print(x_scaled.mean())
print(x_scaled.std())
```

### 1차원 배열 **정규화**

```python
print(x)
```

```python
x_norm = (x-x.min()) /(x.max()-x.min())
print(x_norm)
```

### 2차원 배열에서의 표준화

```python
x  =np.arange(1,13).reshape((4,3)) # 3가지
print(x)
```

```python
mu = x.mean() # 파라미터 axis=None인 경우, 모든 원소(스칼라)들의 평균  1 방법
print(mu)
```

```python
sigma= x.std() # 파라미터 axis=None인 경우, 모든 원소(스칼라)들의 평균
print(sigma)
```

```python
x_scaled = (x-mu)/sigma
print(x_scaled)
```

```python
# axis=0 축 방향으로 평균을 계산  2 방법
mu = x.mean(axis=0)           #keepdims=True
print(mu)  # mu shape:(3, )   (1,3)
```

```python
# axis=0 축 방향으로 표준편차들을 계산
sigma= x.std(axis=0)
print(sigma)  # sigma shape(3, )
```

```python
# axis=0 축 방향으로 값들을 표준화
x_scaled = (x - mu)/sigma
print(x_scaled)
```

```python
# axis=1 축 방향으로 평균들을 계산 3 방법
mu=x.mean(axis=1,keepdims=True)
print(mu) #shape(4, ) >> shape(4,1)
```

```python
# mu.reshape((4,1))
```

```python
x-mu
```

```python
# axis=1 축 방향으로 표준편차들을 계산
sigma=x.std(axis=1,keepdims=True)
print(sigma) #shape(4,1)
```

```python
x_scaled=(x-mu)/sigma
print(x_scaled)
```

```python
# 2차원 배열의 정규화 axis=0,1,X
x_norm=(x-x.min())/(x.max()-x.min())
x_norm
# x_norm0=
# x_norm1=
```

```python
x_norm0=(x-x.min(axis=0))/(x.max(axis=0)-x.min(axis=0))
x_norm0
```

```python
x_norm1=(x-x.min(keepdims=True,axis=1))/(x.max(keepdims=True,axis=1)-x.min(keepdims=True,axis=1))
x_norm1
```
