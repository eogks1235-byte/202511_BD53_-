```python
class Point:
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y
    #self 메서드를호출한객체
  def move(self,dx,dy):
    """x좌표를 dx만큼, y좌표를 dy만큼이동 """
    self.x+=dx
    self.y+=dy
    return self #pt2.move=self(**pt2**).move 연쇄호출 가능하게해줌 나중에 많이사용한다고함
  def __repr__(self):
    return f'Point(x={self.x},y={self.y})'
```

```python
pt1=Point() #x좌표가 0이고, y좌표가 0인 Point객체를생성
print(pt1)
pt1
pt1.move(dx=1,dy=0) #pt1을 x축의 방향으로 1만큼이동
print(pt1)
pt1.move(0,-1) #pt1을 y축의 방향을 -1만큼 이동
print(pt1)
# #16진수
```

```python
pt2=Point(1,1)  # 객체생성문제 x
pt2.move(1,0).move(0,1) #AttributeError: 'NoneType' object has no attribute 'move' /////위에 return self 가 없었을때 실행하면 앞쪽의 오류가 나타난다
print(pt2)
```

```python
#  def area(self):
#도형은 점을 가지고 있다 Shape HAS-A point
class Shape:
  def __init__(self,x=0,y=0):
    """Point 타입 객체를 생성하기 위한 x좌표와 y좌표를 **아규먼트**(앞에사용한)로 전달받음"""
    self.point=Point(x,y)
    """Point 타입 객체를 아규먼트로 전달바듬"""
    #self.point=point #sh=Shape(pt1)포인트객체선택호출 (self,point)
  def __repr__(self):
    return f'Shape(point={self.point})'
  def move(self,dx=0,dy=0):#만약 무브함수를 안썻으면 sh.point.move(,)
    self.point.move(dx,dy)
    return self #move메서드를 호출한 객체(self)를 리턴함
  # def area(self):
  #   pass
  def area(self):
    raise NotImplementedError('넓이 계산 메서드를 구현해야함')
  def perimeter(self):
    raise NotImplementedError('둘레 길이 계산 메서드를 구현해야함')
```

```python
sh=Shape()
print(sh)
sh.move(1,0).move(0,1)
print(sh)
print(sh.area())
print(sh.perimeter())
```

```python
#Shape를 상속하는 Circle클래스
class Circle(Shape):
  def __init__(self, x=0,y=0,radius=0):
    #super().class 객체를 생성(초기화)한다.
    super().__init__(x,y) #상위 클래스 객체 생성(초기화)--Shape클래스의 __init__
    #self.point=Point(x,y) 상속을 안받았을때
    self.radius=radius

  def move(self, dx,dy):
    self.point.move(dx,dy)
  #__str__ 상속이된다
  def __str__(self):#override
    return f'Circle(center={self.point},radius={self.radius})'
#                           self.point=Point(x,y)

  def area(self):
    return 3.14*(self.radius**2)
  def perimeter(self):
    return 3.14*(self.radius*2)
```

```python
c1=Circle()
print(c1)
c1.move(1,2)
print(c1)
c2=Circle(radius=10)
print(f'넓이 ={c2.area()},둘레={c2.perimeter()}')
```

```python
#Shape 클래스를 상속하는  Rectangle클래스
class Rectangle(Shape):
  def __init__(self,corner_x=0,corner_y=0,width=0,height=0):
    super().__init__(corner_x,corner_y)#--Shape클래스의 __init__
    self.width=width
    self.height=height
  def __repr__(self):
    return f'Rectangle(corner={self.point},width={self.width},height={self.height})'
  def area(self):
    return self.width*self.height
  def perimeter(self):
    return 2*(self.width+self.height)
```

```python
rect=Rectangle(width=4,height=3)
print(rect)
print(f'직사각형 넓이={rect.area()}')
print(f'직사각형 둘레={rect.perimeter()}')
rect.move(-1,-1)
```

```python
# Rectangle클래스를 상속하는 Squere 클래스
class Square(Rectangle):
  def __init__(self, corner_x=0, corner_y=0,side=0):
    super().__init__(corner_x,corner_y,side,side)
  def __repr__(self):
    return f'Square(corner={self.point},side={self.width})'
```

```python
#T
square=Square(side=5)
print(square)
print(f'정사각형넓이={square.area()}')
print(f'정사각형둘레={square.perimeter()}')
```
