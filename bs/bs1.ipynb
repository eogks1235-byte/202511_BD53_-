BeautifulSoup 라이브러리를 사용해서 Google Drive에 저장된 HTML문서를 분석.

# BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
web2_html='/content/drive/MyDrive/Colab Notebooks/lab_python/dataanalyst/lab_ml/web2.html'
```

```python
# 1. HTML 문서를 오픈.
html =open(file=web2_html) #mode='rt: 텍스트 읽기 모드. 기본값
```

```python
# 2. HTML 문서를 분석(parsing)할 수 있는 BeautifulSoup 객체 생성
soup = BeautifulSoup(markup=html, features='html5lib') # features='html.parser' 기본값.
# markup 파라미터: 오픈된 HTML 파일(스트림) 객체.
# features 파라미터: HTML 문서를 분석(parsing)하는 라이브러리. 기본값은 'html.parser'.
```

```python
# 3. 생성된 BeautifulSoup 객체의 메서드를 사용해서 HTML 문서의 여러가지 요소들을 찾음.
# bs.find(name='태그이름): HTML 문서에서 가장 처음 나오는 태그 요소를 찾아서 리턴.
element =soup.find(name='h1')
print(element)
```

```python
element =soup.find(name='div')
print(element)
```

```python
element=soup.find(name='a')
print(element)
# HTML 문서의 a 태그 요소들 중에서 첫번쨰 a 요소만 리턴.
```

```python
print(element.text) # a 태그 요소의 컨텐트(내용) 텍스트
print(element.get(key='href')) # a 태그의 href 속성에 설정된 값
```

```python
#bs.find_all(name='테그이름'):같은 태그를 갖는 모든 요소들의 리스트를 리턴.
links=soup.find_all(name='a')
#print(links) # find_all 메서드의 리턴 값은 리스트

for i in links:
  print(i)
  print(i.text)
  print(i.get(key='href'))
  print(i.text, i.get(key='href'))
  print()
```

```python
# 열려져 있는 파일 닫기
html.close()
```

```python
# web2.html 파일을 텍스트 읽기 모드로 오픈
with open(file=web2_html, mode='rt') as f:
  # BeautifulSoup 객체 생성
  bs =BeautifulSoup(markup=f, features='html5lib')

  # div 태그 중에서 class(class) 속성 값이 c1(value)인 첫번째 요소 찾기
  element= bs.find(name='div', attrs={'class':'c1'})
  print(element)

  print('-'*50)
  # div 태그 중에서 class 속성 값이 c1인 모든 요소 찾기(전부)
  elements =bs.find_all(name='div',attrs={'class':'c2'})
  print(elements)
  print()
  for x in elements:
    # print()
    print(x.text)

    # find , find_all
```

HTML 요소(element)를 찾는 방법(CSS Selector):
* `tag_name`
* `.class_name`
* `#id`
* `parent > child`
* `ancestor descendent(사이 공백)`
* `selector1, selector2, ...`
* `selector:pseudo-selector`

```python
with open(file=web2_html) as f:
  bs=BeautifulSoup(markup=f, features='html5lib')
  # tag 이름은 a이고, class 속성값이 menu_link인 모든 요소 찾기
  elements=bs.find_all(name='a',attrs={'class':'menu_link'})
  print(elements)
  print()
  elements=bs.select(selector='a.menu_link')
  print(elements)
  print()
  # ㅇㅈ 없어
  for x in elements:
    print()
    print(x)

    print(x.text)
    print(x.text,x.get(key='href'))
```

```python
# 태그 이름이 section이고 id 속성 값이 sec1 또는 sec2인 요소들 찾기
elements=bs.select(selector='section#sec1, section#sec2')
for x in elements:
  print(x)
```

```python
# 태그 이름이 section이고 아이디가 sec1인 요소를 부모로 갖는 자식요소 p를 찾기
elements =bs.select('section#sec1 > p')
print(elements)
for x in elements:
  print(x)
```

```python
# 태그 이름이 section이고 아이디가 sec1인 요소를 조상으로 갖는 자손 요소 p를 찾기 (전부)
elements =bs.select('section#sec1  p')
print(elements)
for x in elements:
  print(x)
```

```python
# pseudo-selector활용: 태그 이름이 li인 요소들 중에서 세번째 요소만 찾기
link =bs.select('li:nth-of-type(1)')
print(link)
```
