# Scatter plot(산점도)

두 변수 간의 상관관계를 시각화.

## `scatterplot()`

```python
# tip ~ total_bill 산점도
sns.scatterplot(data=tips, x='total_bill',y='tip',
                s=100, c='IndianRed', marker='*',alpha=0.764864856)
#> plt.scatter() 함수의 파라미터들을 그대로 사용할 수 있음.
#> s=마커 크기. c=마커의 색깔. marker=마커 종류. alpha=불투명도
plt.grid()
plt.show()
```

## `regplot()`

regression plot: 산점도 + 선형 회귀(linear regression) 직선.

```python
sns.regplot(data=tips, x='total_bill',y='tip',marker='x',
            scatter_kws={'alpha':0.378895,'s':100},line_kws={'color':'IndianRed'})
#> scatter_kws: 산점도의 설정을 변경하기 위한 dictionary.
#> line_kws: 선형회귀 직선의 설정을 변경하기 위한 dictionary.
plt.grid()
plt.show()
```

## `jointplot()`

산점도 + 히스토그램

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show() # 상단 total bill histo분포도     우측 tip histo분포도
```

# Category plot(`catplot()`)

카테고리 별로 Axes 객체를 생성하고, 변수의 통계값을 모든 Axes에 같은 종류의 그래프로 시각화.


*   요일(x축)별 팁(y축)의 평균 막대 그래프
*   성별(Axes 행)과 흡연여부(Axes 열)별로 Axes들을 나눔


```python
sns.catplot(data=tips,x='day',y='tip',row='sex',col='smoker',kind='bar', errorbar=None)
#> estimator 파라미터의 기본값은 'mean'.
plt.show()
```

```python
tips.groupby(by=['sex','day','smoker'],observed=True).tip.mean()
```

## Ex.1

*   흡연여부(Axes 행), 시간(Axes 열)으로 Axes들을 나눠서, 요일별 팁의 막대 그래프.
*   groupby 메서드와 통계 함수를 사용해서 값 계산.

```python
sns.catplot(data=tips, x='day',y='tip',row='smoker',col='time',kind='bar')
plt.grid()
plt.show()
```

```python
tips.groupby(by=['sex','time','day'],observed=True).tip.mean()
```

## Ex.2

*   흡연여부(Axes 행), 시간(Axes 열)으로 Axes를 나눠서, 요일별 성별 팁의 평균 막대 그래프.
*   groupby를 사용한 값 계산.

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',row='smoker',col='time',kind='bar')
plt.show()
```

```python
tips.groupby(by=['sex','time','smoker','day'],observed=True).tip.mean()
```

## Ex.3

*   흡연여부(Axes 행)와 시간(Axes 열)으로 Axes를 나눠서, 요일별 팁의 box plot.
*   groupby,describe 메서드를 사용해서 기술 통계량을 출력.

```python
g=sns.catplot(data=tips, x='day',y='tip',row='smoker',col='time',kind='box')
# print(g.axes) #> Axes들을 갖는 2차원 배열.
for row in g.axes:
  for x in row:
    x.grid()

# 모든 Axes의 레이블 표시.
g.tick_params(axis='x',labelbottom=True)
g.tick_params(axis='y',labelleft=True)
plt.show()
```

```python
tips.groupby(by=['day','smoker','time'],observed=True).tip.mean()
```

```python
tips.groupby(by=['day','smoker','time'],observed=True).tip.describe() # 아하
```

```python
# 요일별 흡연여부별 팁의 box plot
figure=px.box(data_frame=tips, x='day',y='tip',color='smoker')
figure.show()
```

## scatter plot

```python
# tip ~ total_bill
figure=px.scatter(data_frame=tips, y='tip',x='total_bill')
figure.show()
```

```python
# tip ~ total_bill 산점도.
# 성별로 색깔을 다르게, 요일별로 마커 모양을 다르게, 인원수별로 마커 크기를 다르게
figure=px.scatter(data_frame=tips, y='tip',x='total_bill',color='sex',size='size',symbol='day')
figure.show()
```

## Histogram

```python
figure = px.histogram(data_frame=tips, x='total_bill',nbins=10,
                      template='simple_white')
figure.show()
```

## Facetting

Figure 안에 행과 열을 구분해서 Axes들을 여러개 생성하고 그래프를 시각화.

```python
# 흡연여부와 시간으로 Axes를 나눠서 tip ~ total_bill 산점도.
figure=px.scatter(data_frame=tips, y='tip',x='total_bill', facet_row='smoker',facet_col='time', color='sex')
figure.show()
```
