# Python `list` 클래스

*   여러개의 값들을 저장할 수 있는 데이터 타입.
*   인덱스를 기반으로 한 데이터 타입.
    *   indexing: 인덱스를 사용해서 리스트의 특정 위치의 값을 참조.(예) `list[0]`
    *   slicing: [start,end) 인덱스 범위의 리스트를 잘라냄. (예) `list[0:5]`
*   산술 연산
    *   `list + list`: concatenate. 리스트 이어붙이기.
    *   `list * int, int * list`: replicate. 리스트의 원소들을 정수만큼 복사.

```python
numbers1=[1,2,3]
numbers2=[11,22,33]
```

```python
print(numbers1) #대괄호 쉼표구분
```

```python
numbers2
```

```python
# indexing
print(numbers1[0]) #리스트의 첫번째 원소
print(numbers1[-1]) #리스트의 마지막 원소
```

```python
#slicing
print(numbers2[0:2]) # 리스트에서 첫 2개 원소들을 잘라내기
print(numbers2[-2:]) # 리스트에서 마지막 2개 원소들을 잘라내기
```

```python
numbers1 +numbers2
```

```python
numbers1 * 3
```

```python
# numbers1과 numbers2의 같은 인덱스에 있는 원소들끼리 덧셈 --> [12,24,36]
asd=[x+y for x,y in zip(numbers1, numbers2)]
print(asd)
```

```python
# numbers1의 모든 원소에 3을 곱함
[x*3 for x in numbers1]
```

```python
# numbers1과 numbers2의 같은 인덱스에 있는 원소들끼리 곱셈 --> [11,44,99]
[x*y for x,y in zip(numbers1,numbers2)]
```

# NumPy(Numerical Python)

*   다차원배열(n-dimensional array)을 쉽고 빠르게 연산(`+,-,*,/,...`)하기 위한 패키지.
*   numpy 패키지의 대부분의 기능들(연산자, 함수, 메서드 등)은 반복문의 기능을 가지고 있음.
    *   같은 인덱스의 원소들끼리(element-wise) 연산하는 기능을 가지고 있음.
*   `numpy.ndarray` 클래스: numpy 패키지의 가장 기본이 되는 클래스.
    

```python
num_array1 = np.array([1,2,3,4,5])
```

```python
type(num_array1)
```

```python
print(num_array1) #  ,가없다 ' ' 으로출력  ndarray클래스의 __str__() 메서드가 리턴하는 문자열을 출력
```

```python
num_array1    # ndarray 클래스의 __repr__() 메서드가 리턴하는 문자열을 출력
```

```python
num_array2 =np.array([10,20,30,40,50])
```

```python
# element-wise(같은 인덱스의 원소들끼리) 연산
print(num_array1+num_array2) #
print(num_array1-num_array2)
print(num_array1*num_array2)
print(num_array1*3)
print(num_array1/num_array2)
```

# `np.ndarry` 클래스의 속성들(attributes)

```python
print(num_array1)
print('ndim= ',num_array1.ndim) # number of dimension (차원의 개수)
print('size = ', num_array1.size) # 배열의 크기(모든 원소의 개수)
print('shape = ', num_array1.shape) # 배열의 모양(배열의 각 차원의 원소 개수들로 이루어진 튜플)
print('dtype = ', num_array1.dtype) # data type(배열 원소의 데이터 타입)
```

# `np.ndarray` 객체 생성 함수

## `np.array(object)`

*   object를 ndarray로 변환해서 리턴.
*   object: 배열과 비슷한 객체들. list, tuple, ndarray, ...


```python
# 0 이상 10 미만의 정수들로 이루어진 ndarray(배열)만들기
numbers= np.array([x for x in range(0,10)])
print(numbers)
```

## `np.arange(start,end,step)`

*   range를 사용해서 array(배열)을 만들어 주는 함수.
*   파이썬 `range(start, end,step)` 함수 사용법과 같다.
*   1차원 배열(ndarray)를 생성.


```python
# 0 이상 10 미만의 정수들로 이루어진 ndarray(배열)만들기
numbers=np.arange(10)
print(numbers)
```

```python
# 0 이상 10미만의 짝수들로만 이루어진 배열
evens= np.arange(0,10,2)
print(evens)
```

## reshape

*   `np.reshape(array, newshape)` 함수(function)
*   `np.ndarray.reshape(newshape)` 메서드(method)


```python
array = np.arange(12)
print(array)
```

```python
# reshape 함수
array2= np.reshape(array, (3,4))#모양은 tuple로
print(array2)
```

```python
# reshape 메서드
print(array)
array3=array.reshape((4,3))
print(array3)
```

```python
# 함수(메서드) 연쇄 호출
array =np.arange(15).reshape((3,5))
print(array)
```

## 난수 배열

### `np.random.rand(d0, d1, d2, d3, ...)`

*   [0,1) 범위의 **균등분포(uniform distribution)**를 따르는 난수들의 배열을 생성.
*   아규먼트 (d0, d1, ...)들은 각 차원의 원소 개수.


```python
array=np.random.rand(5)
print(array)
```

```python
array=np.random.rand(3,4)
print(array)
```

### ` np.random.randn(d0, d1, d2, ...)`

*   평균이 0이고 표준편차가 1인 **표준 정규 분포(standard normal distridution)**를 따르는 난수들의 배열을 생성.
*   아규먼트(d0, d1, d2, ...)는 각 차원의 원소 개수.

```python
array =np.random.randn(10)
print(array)
```

```python
array=np.random.randn(3,5)
print(array)
```

### `np.random.randint(low, high, size, dtype)`

*   [low, high) 범위의 정수 난수들을 원소로 하는 배열을 생성.
*   파라미터 size: ndarray의 **shape**을 정수 또는 튜플로 전달. (1차원은 원소의개수)         (파라미터 size는 shape이다)
*   파라미터 dtype: int8, int16, int32, int64,...

```python
array=np.random.randint(0,10,size=5)  #파라미터 size에 정수를 전달 >> 1차원 배열
print(array)
```

```python
array=np.random.randint(0,10,size=(5,))  # shape이 (5, )인 1차원 배열
print(array)
```

```python
array= np.random.randint(0,10,size=(4,5))  #shape이 (4,5)인 2차원 배열
print(array)
```

# 통계 함수/메서드


*   합계 평균 분산 최대 최소 표준편차 중앙값(X함수만있다), ...
*   함수: `np.function_name(ndarray)`
*   메서드: `np.ndarray.method_name()`

```python
array =np.random.randint(100,size=10)   # 0 이상 100 미만 10개의 정수 난수를 갖는 배열
print(array)
```

```python
print('합계 : ',np.sum(array), array.sum()) # 배열을 아규먼트로 /// 배열안에서 합계
```

```python
print('평균 : ',np.mean(array), array.mean()) #numpy의 함수 /// 만든 객체의 기능
```

```python
print('분산 : ',np.var(array), array.var())
```

```python
print('표준편차 : ',np.std(array), array.std())
```

```python
print('최댓값 : ',np.max(array), array.max())
```

```python
print('최솟값 : ',np.min(array),array.min())
```

```python
print('중앙값 : ',np.median(array)) # 중앙값 메서드는 없다
```

## 2차원 배열에서의 통계 함수/메서드

```python
np.random.seed(42) # seed가 같으면 값 일정하다 /// 실행할 때마다 같은 난수들을 생성하기 위해서.
array=np.random.randint(10,size=(5,4))
print(array)   # 축  axis 세로 0 가로 1
```

```python
np.sum(array) # 2차원 배열의 모든 원소들의 합계
# print(np.sum(array))
```

```python
print(np.sum(array))
```

```python
array.sum()
```

```python
np.sum(array, axis=0) #세로길이합 (axis=0 축을 따라서 계싼한 원소들의 합계)
```

```python
array.sum(axis=0)
```

```python
np.sum(array,axis=1) #가로길이합 (axis=1 축을 따라서 계싼한 원소들의 합계)
```

```python
array.sum(axis=1)
```

```python
array.sum(axis=0,keepdims=True) # 2차원 배열 세로길이합 (1,4)shape  False는 기본 axis sum과 동일한 결과가나옴
```

```python
array.sum(axis=1,keepdims=True) # 2차원 배열 가로길이합 (5,1)shape
```
