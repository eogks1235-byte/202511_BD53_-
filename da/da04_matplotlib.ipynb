# `matplotlib.pyplot` 모듈을 사용한 데이터 시각화

```python
# 1. Figure, Axes 객체 생성.
fig, axis = plt.subplots() # nrows, ncols 파라미터 기본값은 1.

# 2. 그래프 그림.
axis.hist(x=tips.tip)

# 3. 그래프 옵션 설정.
axis.set_title('Histogram')
axis.set_xlabel('tip')
axis.set_ylabel('count')

# 4. 그래프 보여주기.
plt.show()
```

# Scatter plot(산점도)

*   두 변수 간의 상관관계를 알고 싶을 때 사용.
*   `y ~ x` 관계 시각화.
    *   y: 관심변수, 종속변수
    *   x: 독립변수

```python
# tip(팁 y) ~ total_bill(영수증 총액 x) 관계
plt.scatter(x=tips.total_bill, y=tips.tip,  # x: x축에 사용할 변수(컬럼), y: y축에 사용할 변수
            alpha=0.3,# alpha: 불투명도(0=투명, 1=불투명)  ///  곂쳐진정도 진하게 불투명도
            s=50,  # s:size 마커(점)의 크기  ///  점의크기
            c='IndianRed', # c: color 머커(점)의 색깔 HTML color name: W3홈페이지
            marker='+', # marker: 마커(점)의 모양 (v o r * + ^ s ...)
            )
plt.xlabel('total bill')
plt.ylabel('tip')
plt.grid()
plt.show()
```

*    tip ~ total_bill 상관관계를 산점도로 표현.
*    성별에 따라서 마커의 색깔을 다르게 표현.

```python
male_df =tips[tips.sex=='Male']  # 성별이 남성인 데이터 프레임
female_df =tips[tips.sex=='Female'] # 성별이 여성인 데이터 프레임

plt.scatter(x=male_df.total_bill, y=male_df.tip,
            alpha=0.3, c='DodgerBlue', label='male')
plt.scatter(x=female_df.total_bill,y=female_df.tip,
            alpha=0.3, c='IndianRed', label='female')
plt.xlabel('total bill')
plt.ylabel('tip')
plt.grid()
plt.legend() # 범례표시 (scatter 함수의 label 파라미터 값을 표시).
plt.show()
```

*   tip ~ total_bill의 산점도
*   성별에 따라서 마커의 색깔을 다르게 표현
*   흡연여부에 따라서 마커의 모양을 다르게 표현

# Line plot(선 그래프)

주로 시계열 데이터(주식, 환율, 부동산, 지지도, 미세먼지농도, ...)를 시각화.

```python
# 선 그래프를 그리기 위한 가상의 데이터
np.random.seed(1) # 실행할 때마다 같은 난수들이 만들어 지도록 하기 위해서.
df = pd.DataFrame(data={
    'day':np.arange(1,11),
    'price':np.random.rand(10),

})
df
```

`plt.plt([x], y, [fmt])`

*   x: x좌표로 사용할 변수. *생략 가능.* 생략했을 때 인덱스가 x좌표로 사용됨.
*   y: y좌표로 사용할 변수. **필수.**
*   fmt: 포멧 문자열. color(선과 마커의 색깔), marker(마커 모양), linestyle(선 스타일) 정보를 문자열로 표현. *생략가능.* 생략했을 때는 마커 없이 직선으로 표시.
    *   color: r, g, b, k, ... : (red, green, blue, black, ...)
    *   marker: o, s, ^, v, *, +, ...(생략가능)
    *   linestyle: -(직선, soild line), --(dashed line), :(점선, dotted line), ...
    *   (예) 'ro-','bo:'

```python
# plt.plot([x],y,[fmt]) []생략가능
plt.plot(df.day,df.price, 'bo:') #[fmt] 색 마커 선 지정
plt.grid()
plt.show()
```

```python
plt.plot(df.day, df.price,color='IndianRed',marker='+',linestyle='dashdot')
#'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.xlabel('day')
plt.ylabel('price')
plt.show()
```
