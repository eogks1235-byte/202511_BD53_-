# CSV 파일의 첫번째 행부터 데이터인 경우

컬럼 이름이 파일에 없는 경우

```python
ex2_path='https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex2.csv'
```

```python
ex2_df=pd.read_csv(ex2_path, names=['col1','col2','col3','col4','col5'])
```

```python
ex2_df # 컬럼으로 올라가 버림 names를만들면 거기적힌게 컬럼이됨
```

```python
ex2_df.info()
```

`read_csv()` 함수의 names 파라미터는 DataFrame의 컬럼 이름을 설정할 때 사용.

*   CSV 파일에 컬럼 이름이 없는 경우
*   CSV 파일에 컬럼 이름이 있지만, 컬럼 이름을 재지정할 때

```python
ex1_df=pd.read_csv(ex1_path, names=['korean','english','math','science','message'],
                   header=0) # 파일의 첫번째 라인은 데이터가 아니라 헤더임을 알려줌.
```

```python
ex1_df
```

```python
ex1_df.info()
```

# 값들이 공백으로 구분된 텍스트 파일(txt, csv, tsv, ...)

```python
ex3_path='https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex3.txt'
```

```python
ex3_df=pd.read_csv(ex3_path,sep='\\s+')# 여러개의공백으로 나
# ex3_df=pd.read_csv(ex3_path,sep='\t')
```

```python
ex3_df
```

```python
ex3_df.info()
```

ex3.txt 파일은 컬럼 이름과 행 인덱스를 가지고 있는 형식.

```python
ex3_df.loc['aaa']
```

```python
ex3_df.iloc[0]
```

```python
ex3_df.loc['aaa':'bbb']
```

```python
ex3_df.iloc[0:2]
```

# 컬럼 헤더(이름)와 데이터 이외의 주석등을 포함하는 파일

```python
ex4_path='https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex4.csv'
```

```python
ex4_df=pd.read_csv(ex4_path,skiprows=[0,2,3])
ex4_df
```

*   csv(txt) 파일 >>> DataFrame: `pq.read_csv()` 함수사용.
*   DataFrame >>> csv 파일 저장: `pd.DataFrame.to_csv()` 메서드 사용.

```python
# ex3_df를 ex3.csv 파일로 저장
ex3_df.to_csv('ex3.csv')
```

```python
# ex4_df를 ex4.csv 파일로 저장
ex4_df.to_csv('ex4.csv',index=False)
```
