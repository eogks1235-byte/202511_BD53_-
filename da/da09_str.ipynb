### 가장 많이 등장하는 영화 장르 상위/하위 10개

```python
genres=[] # 장르들을 저장하기 위한 빈 리스트
for x in mfdf.genres:
  genres.extend(x.split('|')) # append 문자열자체 [['Animation', "Children's", 'Comedy']]

print(genres)  # ['Animation', "Children's", 'Comedy']
```

```python
# 리스트를 Series로 변환 >> Series의 메서드들을 사용하기 위해서
genres_sr=pd.Series(genres)
genres_sr
```

```python
genre_count = genres_sr.value_counts()
genre_count
```

```python
genres_sr.unique()
```

```python
genres_sr.unique().size
```

```python
# 가장 많이 등장하는 상위 10개 장르

top_10=genre_count.head(n=10)
top_10
```

```python
# sns.barplot(data=top_10) # 세로 막대그래프
sns.barplot(x=top_10.values,y=top_10.index) # 가로 막대그래프
plt.show()
```

```python
# 하위 10개 장르
bottom10= genre_count.tail(n=10)
bottom10
```

```python
sns.barplot(x=bottom10.values,y=bottom10.index)
plt.show()
```
