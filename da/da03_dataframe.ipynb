# TITANIC 데이터셋

```python
titanic= sns.load_dataset('titanic')
```

```python
titanic.head()
```

titanic 데이터 프레임의 컬럼들:

*   survived: 생존여부(0=사망,1=생존) 중복되는데이터 삭제 하는게 좋다
      * alive 컬럼과 같은내용(no=사망, yes=생존)
*   pclass:   객실등급
      * class 컬럼과 같은내용(1=First, 2=Second, 3=Third)
*   sex:      성별(male/female)
*   who:      성인남성(man), 성인여성(woman), 어린이(child)
*   adult_male: 성인 남성 여부(True, False)
*   age:      나이
*   sibsp:    동승한 형제자매(sibling) 또는 배우자(spouses) 수
*   parch:    동승한 부모(parents) 또는 자녀(children) 수
*   alone:    동승자 여부(True=동승자 없음(혼자), False=동승자 있음(혼자가아님))
*   fare:     운임(탑승권 가격)
