### Multi-level index에서 indexing, slicing

nlevels(인덱스의 계층 개수)가 2개 이상인 multi-level 인덱스를 loc 속성에서 사용할 때,

*   첫번째 계층의 인덱스만 가지고 indexing과 slicing을 할 수 있음.
*   두번째 계층의 인덱스만 가지고는 indexing과 slicing을 할 수 없다.
*   튜플 형태의 인덱스로 indexing/slicing을 할 수 있음.


```python
s
```

```python
s.loc['m'] # 첫번째 계층(level)의 인덱스로 indexing.
```

```python
s.loc['f']
```

```python
# s.loc['a'] # 이 방법은 가장 바깥에것(1레벨)만 찾을수있다 레벨 1  abc는 레벨 2 //  KeyError-'a'는 첫번째 계층(level)의 인덱스가 아니기 때문이다.
s.loc[('m','a')] #튜플 형태의 인덱스를 사용한 indexing.
```

```python
s.loc[[('m','a'),('f','a')]]    # loc인덱스 > 리스트 > 인덱스 fancy
```

```python
s_idx_sorted=s.sort_index()
s_idx_sorted
```

```python
s_idx_sorted.loc['f':'m'] # 정렬되니까 된다  slicing 은 인덱스들이 정렬된 상태일 때 가능한것
```

```python
s_idx_sorted.loc[('f','a'):('m','a')]
```

*   `pd.Series.sort_index()`: 인덱스를 정렬(한 새로운 Series 객체를 리턴. )
*   `pd.Series.sort_values()`: 값들을 정렬
*   `inplace` 파라미터:
      *   `inplace=True`: 원본 Series를 정렬된 상태로 변경.
      *   `inplace=False`: 원본은 수정되지 않고, 정렬된 새로운 Series를 객체를 리턴. 기본값은 False다

```python
np.random.seed(42)
s= pd.Series(data=np.random.rand(5), index=['b','a','z','d','c'])
s
```

```python
s.sort_values() # 값 정렬 inplace = False 기본값
```

```python
s # 원래 순서로
```

```python
s.sort_values(inplace=True) # inplace=True 원본변경, 리턴값은 없음
s
```

```python
s # 원본 Series가 값들 오름차순 정렬된 상태
```

```python
s.sort_index() # inplace=False: 기본값. 인덱스들이 정렬된 Series를 리턴
```

```python
s # 원본은 변경 없음
```

```python
s.sort_index(inplace=True) # inplace=True 원본 Series의 인덱스를 정렬. 리턴 값은 없음.
```

```python
s
```

## index swapping

multi-level index에서 인덱스의 레벨을 바꿈.

```python
np.random.seed(42)
s= pd.Series(data=np.random.randint(100,size=6),index=[['m']*3+['f']*3,['a','b','c']*2])
s
```

```python
s.swaplevel() # 레벨1 과 레벨2 가 바뀜    //copy=True: 기본값.원본은 변경은 없다. copy(X)// inplace=False같은개념
```

```python
s
```

```python
# s.swaplevel(copy=False) False는 없다
```

```python
# nlevels= 3인 인덱스를 갖는 Series
np.random.seed(42)
s=pd.Series(data=np.random.rand(6),
            index=[np.arange(1,7),
                   ['A']*3+['B']*3,
                   ['a','b','c']*2])
s
```

```python
s.index.nlevels
```

```python
s.swaplevel(i=-3,j=-1) #i=-2 ,j=-1: 기본값. 끝에서 두번째 인덱스와 마지막 인덱스를 서로 바꿈.
```

```python
s.swaplevel(i=0,j=1) # 첫번째 인덱스와 두번째 인덱스를 바꿈
```

## DataFrame 컬럼 <---> DataFrame 인덱스

*   `pd.DataFrame.set_index()`: 데이터프레임의 컬럼(들)을 인덱스로 변환.
*   `pd.DataFrame.reset_index()`: 인덱스를 데이터프레임의 컬럼으로 변환.

```python
df
```

## reset_index

```python
df.reset_index() # 기본 동작: 모든 level의 인덱스를 컬럼으로 변환.
```

```python
# 인덱스를 컬럼으로 변환할 때 사용할 컬럼 이름들을 설정.
df.reset_index(names=['day','time'])
```

```python
df.reset_index(level=0, names='day') # level=0만 컬럼으로넣는다  한개일때 names=['day'], names='day' 둘다가능
```

```python
df.reset_index(level=1,names=[None,'time']) # 가장마지막 인덱스 두번째 위치에있던거라서 앞에 대체할만한 뭔가가 필요했다
```

## `set_index`

```python
# class = 1 인 데이터 선택
exam[exam['class']==1]
```

```python
exam_cls = exam.set_index(keys=['class']) #s 는 []인 경우가 많다
exam_cls
```

```python
exam_cls.loc[1]
```

```python
exam.set_index(keys=['class','id'])
```

```python
exam.set_index(keys=['id','class']) # loc(행,열)
```
