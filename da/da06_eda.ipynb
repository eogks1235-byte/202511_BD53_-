# mpg 데이터셋

* seaborn 패키지의 예제 데이터 셋 'mpg'를 로드해서 DataFrame 생성

* DataFrame 확인 - 출력, 요약 정보, 숫자 타입 기술 통계량, 카테고리 타입 빈도수

* 산점도

  * mpg(연비) ~ displacement(배기량)
  * 2x2 모양의 subplot을 생성
    * mpg ~ cylinders(실린더 개수)
    * mpg ~ horsepower(마력)
    * mpg ~ weight(무게)
    * mpg ~ acceleration(가속도)

* 빈도수 막대 그래프

  * cyliners 빈도수
  * origin(자동차 생산 지역) 빈도수
  * model_year(자동차 생산 연도) 빈도수
* 선 그래프 가능?
  * origin 별, cylinders 별 빈도수
* 통계량 막대 그래프 - 2개의 subplot 생성

  * cylinders 별 mpg의 중앙값(median)
  * origin 별 mpg의 중앙값
* Box plot

  * 3개의 subplot을 생성
    * mpg, displacement, weight box plot
  * origin 별 mpg box plot
* Histogram

  * 4개의 subplot을 생성
    * mpg, displacement, weight, model_year histogram
* Pair plot

mpg 데이터 프레임의 컬럼:

*   mpg: Mile per Gallon. 자동차 연비.

## scatterplot (산점도)

```python
# mpg(연비)관심변수(종속) ~ displacement(배기량)독립변수
sns.scatterplot(data=mpg, y='mpg', x='displacement')
plt.show()
```

```python
sns.regplot(data=mpg, y='mpg',x='displacement',scatter_kws={'alpha':0.4}, # scatter plot 옵션 설정 - 점들의 투명도
            line_kws={'color': 'IndianRed'})# line plot 옵션 설정 - 선 색 옵션
plt.show()
```

```python
# 2x2 Axes를 생성
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(8,8))

# 각각의 Axes에 산점도를 시각화
sns.scatterplot(data=mpg,y='mpg',x='cylinders',ax=axes[0][0],hue='origin')
sns.scatterplot(data=mpg,y='mpg',x='horsepower',ax=axes[0][1],hue='origin')
sns.scatterplot(data=mpg,y='mpg',x='weight',ax=axes[1][0],hue='origin')
sns.scatterplot(data=mpg,y='mpg',x='acceleration',ax=axes[1][1],hue='origin')

plt.show()
```

```python
sns.regplot(data=mpg,y='mpg',x='acceleration')

plt.show()
```

상관계수(correlation coefficient)

*   두 변수의 상관관계를 수치로 표현한 값.
*   Pearson, Spearman, Kendall 상관계수.
*   -1 이상 1 이하의 값을 찾음.
*   상관계수가 양수인 경우, 양의 상관관계를 갖는다고 말함.
*   상관계수가 음수인 경우, 음의 상관관계를 갖는다고 말함.
*   상관계수가 +1/-1에 가까울 수록 상관관계가 크다고 말함. (직선)
*   상관계수가 0에 가까울 수록 상관관계가 작다고 말함. (원형)

```python
# corr 상관계수 pearson
mpg[['mpg','displacement','cylinders','horsepower','weight','acceleration']].corr()
```

```python
# corr 상관계수 spearman
mpg[['mpg','displacement','cylinders','horsepower','weight','acceleration']].corr(method='spearman')
```

```python
# corr 상관계수 kendall
mpg[['mpg','displacement','cylinders','horsepower','weight','acceleration']].corr(method='kendall')
```
