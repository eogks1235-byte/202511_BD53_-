## DataFrame에서 column 선택하는 방법

*   `data_frame['column_index(name)']`
    *   항상 사용이 가능하다.
*   `data_frame.column_name`
    *   컬럼의 이름이 변수 이름 규칙에 맞지 않는 경우에는 사용할 수 없음.
        *   변수 이름은 영문자, 숫자, _(underscore)만 사용 가능.
        *   변수 이름은 숫자로 시작할 수 없음.
        *   파이썬 키워드(if, for, while, def, class, ...)들은 변수 이름으로 사용할 수 없음.
    *   컬럼 이름이 DataFrame 클래스의 속성 또는 메서드 이름과 같은 경우에는 사용할 수 없음.

### boolean indexing

*   조건을 만족하는 행(row)들을 선택하는 방법.
*   `iloc`는 사용할 수 없음.
*   `data_frame.loc[조건식]` 또는 `data_frame[조건식]`

```python
df.loc[[True, True, False,False,True,False]] # iloc도 가능 boolean에서는불가
```

```python
df.city == '서울' # 비교연산
```

```python
df.loc[df.city=='서울'] #True인 값 출력
```

```python
df[df.city=='서울']
```

```python
# 2024년 행만 선택
df[df.year==2024]
```

*   boolean indexing에서는 파이썬 논리 연산자 키워드`and,or,not`을 사용할 수 없음.
*   대신에 `&, |, ~`연산자를 사용함.
*   `&, |`의 피연산자들은 반드시`()`를 사용해야 함.

```python
# year 컬럼의 값이 2023 또는 2024인 행을 선택.
df[(df.year==2023)|(df.year==2024)] # or (X)
```

```python
# 서울 2024년 자료
df[(df.city=='서울')&(df.year==2024)] # and (X) boolean일때 괄호필수
```
